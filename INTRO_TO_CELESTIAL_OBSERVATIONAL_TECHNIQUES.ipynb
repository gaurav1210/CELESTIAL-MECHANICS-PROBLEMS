## **Question No.2**
## **Solution**

**Step 1: Diffraction Limit Formula**

The diffraction limit formula is given by:

$$[
\theta = \frac{1.22 \lambda}{D}
]
$$
where:

{
$(\theta )$ is the angular resolution (diffraction limit) in radians,
$( \lambda )$ is the wavelength of light (in meters),
$( D )$ is the diameter of the telescope's aperture (in meters),
$( 1.22 )$ is a constant for a circular aperture.
}

**Step 2: Hubble Space Telescope Diffraction Limit in the Ultraviolet (300 nm)**

$$
[
\lambda = 300 \, \text{nm} = 300 \times 10^{-9} \, \text{m}
]
$$
$$
[
D_{\text{Hubble}} = 2.4 \, \text{m}
]
$$

Substituting the values into the diffraction limit formula:

$$[
\theta_{\text{Hubble, UV}} = \frac{1.22 \times 300 \times 10^{-9}}{2.4}
]
$$
$$
[
\theta_{\text{Hubble, UV}} = \frac{3.66 \times 10^{-7}}{2.4}
]
$$
$$
[
\theta_{\text{Hubble, UV}} \approx 1.525 \times 10^{-7} \, \text{radians}
]
$$

**To convert this to arcseconds:**
$$
[
1 \, \text{radian} = \frac{180 \times 3600}{\pi} \, \text{arcseconds}
]
$$

So:
$$
[
\theta_{\text{Hubble, UV}} \approx 1.525 \times 10^{-7} \times \frac{180 \times 3600}{\pi}
]
$$
$$
[
\theta_{\text{Hubble, UV}} \approx 0.0313 \, \text{arcseconds}
]
$$

**Step 3: Hubble Space Telescope Diffraction Limit in the Near Infrared (2.0 μm)**

$$
[
\lambda = 2.0 \, \mu\text{m} = 2.0 \times 10^{-6} \, \text{m}
]
$$

Using the same formula:

$$
[
\theta_{\text{Hubble, IR}} = \frac{1.22 \times 2.0 \times 10^{-6}}{2.4}
]
$$
$$
[
\theta_{\text{Hubble, IR}} = \frac{2.44 \times 10^{-6}}{2.4}
]
$$
$$
[
\theta_{\text{Hubble, IR}} \approx 1.017 \times 10^{-6} \, \text{radians}
]
$$

**Converting to arcseconds:**

$$
[
\theta_{\text{Hubble, IR}} \approx 1.017 \times 10^{-6} \times \frac{180 \times 3600}{\pi}
]
$$

$$
[
\theta_{\text{Hubble, IR}} \approx 0.2097 = 0.210 \, \text{arcseconds}
]
$$

**Step 4: James Webb Space Telescope Diffraction Limit at 2.0 μm**

The diameter of the James Webb Space Telescope is:
$$
[
D_{\text{JWST}} = 6.5 \, \text{m}
]
$$

Now we calculate the diffraction limit for JWST at the same infrared wavelength $( \lambda = 2.0 \, \mu\text{m} ):$

$$
[
\theta_{\text{JWST, IR}} = \frac{1.22 \times 2.0 \times 10^{-6}}{6.5}
]
$$
$$
[
\theta_{\text{JWST, IR}} = \frac{2.44 \times 10^{-6}}{6.5}
]
$$
$$
[
\theta_{\text{JWST, IR}} \approx 3.75 \times 10^{-7} \, \text{radians}
]
$$

**Converting to arcseconds:**

$$
[
\theta_{\text{JWST, IR}} \approx 3.75 \times 10^{-7} \times \frac{180 \times 3600}{\pi}
]
$$
$$
[
\theta_{\text{JWST, IR}} \approx 0.077 \, \text{arcseconds}
]
$$

**Step 5: Comparison**

- **Hubble diffraction limit at 300 nm (ultraviolet):** $( 0.0313 \, \text{arcseconds} )$
- **Hubble diffraction limit at 2.0 μm (infrared):** $( 0.210 \, \text{arcseconds} )$
- **JWST diffraction limit at 2.0 μm (infrared):** $( 0.077 \, \text{arcseconds} )$


**Conclusion**

- **Ultraviolet Region:** Hubble excels in the ultraviolet spectrum, where its diffraction limit of 0.0313 arcseconds gives it exceptional angular resolution.

- **Infrared Region:** In the infrared spectrum at 2.0 μm, JWST far outperforms Hubble, achieving a resolution of 0.077 arcseconds, compared to Hubble's 0.210 arcseconds. This is primarily due to the larger aperture of JWST, which is better suited for capturing long-wavelength infrared light with higher resolution.

- **Overall:** JWST's larger mirror makes it a much better instrument for infrared observations, providing a more detailed view of distant objects, while Hubble remains powerful in the ultraviolet regime.

We need to calculate the magnitude threshold for the James Webb Space Telescope (JWST) given that the Hubble Space Telescope (HST) has a magnitude threshold of $( m = 26.0 )$ and the background brightness for the JWST is 1.0 magnitude arcsecond$(^-2)$ fainter than for the HST.

**Step 1: Magnitude Limit Relation**

The magnitude limit \( m \) is proportional to:
$$
[
m \propto 2.5 \log_{10} \left(\frac{D^2}{B}\right)
]
$$
where:

- ( D \) is the aperture diameter and
- ( B \) is the background brightness.


Given that the background brightness \( B \) for JWST is 1.0 mag arcsec$(^-2)$ fainter than for HST, we can express:

$$
[
B_{\text{JWST}} = B_{\text{HST}} - 1.0
]
$$

**Step 2: Ratio of Apertures**

The diameters of the HST and JWST are:

$$[
D_{\text{HST}} = 2.4 \, \text{m}
]
$$
$$
[
D_{\text{JWST}} = 6.5 \, \text{m}
]
$$

The ratio of the squared apertures is:

$$[
\frac{D_{\text{JWST}}^2}{D_{\text{HST}}^2} = \frac{6.5^2}{2.4^2} = \frac{42.25}{5.76} \approx 7.335
]
$$

**Step 3: Adjust Magnitude Threshold**

The magnitude threshold for JWST $( m_{\text{JWST}} )$ can be found using:

$$
[
m_{\text{JWST}} = m_{\text{HST}} + 2.5 \log_{10} \left(\frac{D_{\text{JWST}}^2}{D_{\text{HST}}^2}\right) - (\text{background adjustment})
]
$$

Substituting the values:

$$
[
m_{\text{JWST}} = 26.0 + 2.5 \log_{10}(7.335) - 1.0
]
$$

Calculating $( \log_{10}(7.335) ):$
$$
[
\log_{10}(7.335) \approx 0.865
]
$$

Thus:
$$
[
m_{\text{JWST}} = 26.0 + 2.5 \times 0.865 - 1.0
]
$$
$$
[
m_{\text{JWST}} = 26.0 + 2.1675 - 1.0
]
$$
$$
[
m_{\text{JWST}} \approx 27.1625
]
$$

**Conclusion**

The magnitude threshold for the James Webb Space Telescope (JWST) is approximately $( \mathbf{27.1625} ).$ This is a result of the larger aperture of JWST and the more favorable background brightness.

#### **Part (a): Seeing at a Different Elevation**

**Step 1: Zenith Angle Calculation**

The zenith angle \( z \) is related to the elevation $( \epsilon )$ as:

$$
[
z = 90^\circ - \epsilon
]
$$

At an elevation $( \epsilon = 35^\circ )$, the zenith angle is:

$$
[
z = 90^\circ - 35^\circ = 55^\circ
]
$$

**Step 2: Air Mass Calculation**

The air mass \( X(z) \) is given by:
$$
[
X(z) = \frac{1}{\cos(z)}
]
$$
For $( z = 55^\circ )$, the air mass is:
$$
[
X(55^\circ) = \frac{1}{\cos(55^\circ)} \approx \frac{1}{0.5736} = 1.743
]
$$

**Step 3: Seeing at $( z = 55^\circ )$**

Seeing scales with the air mass as:

$$
[
\text{Seeing}(X) \propto X^{3/5}
]
$$

The seeing at zenith (where \( X = 1 \)) is 0.72 arcseconds, so the seeing at $( X(55^\circ) = 1.743 )$ is:

$$
[
\text{Seeing}(55^\circ) = 0.72 \times (1.743)^{3/5}
]
$$

Now, calculate $( (1.743)^{3/5} ):$

$$
[
(1.743)^{3/5} \approx 1.408
]
$$

Thus, the seeing at 55° is:
$$
[
\text{Seeing}(55^\circ) \approx 0.72 \times 1.408 = 1.014 \, \text{arcseconds}
]
$$

#### **Part (b): Seeing at a Different Wavelength**


**Step 4: Wavelength Scaling**


Seeing also scales with the wavelength $( \lambda )$ as:

$$
[
\text{Seeing}(\lambda) \propto \lambda^{-1/5}
]
$$

Given that the seeing at 400 nm is 0.72 arcseconds, the seeing at 2.2 μm (or 2200 nm) is:

$$
[
\text{Seeing}(2.2 \, \mu\text{m}) = 0.72 \times \left( \frac{400 \, \text{nm}}{2200 \, \text{nm}} \right)^{1/5}
]
$$

First, calculate the wavelength ratio:
$$
[
\frac{400 \, \text{nm}}{2200 \, \text{nm}} = 0.1818
]
$$

Now, raise this ratio to the power of \( 1/5 \):

$$
[
(0.1818)^{1/5} \approx 0.668
]
$$

Thus, the seeing at 2.2 μm is:

$$
[
\text{Seeing}(2.2 \, \mu\text{m}) = 0.72 \times 0.668 = 0.481 \, \text{arcseconds}
]
$$

####  **Final Answer:**

- The seeing at an elevation of 35 degrees is approximately **1.014 arcseconds**.
- The seeing at 2.2 μm for the same star is approximately **0.481 arcseconds**.

A MOS (Metal-Oxide-Semiconductor) capacitor is a fundamental component in modern electronics, particularly in MOSFET transistors. Here's how it works with an n-type semiconductor:

**Structure:**

- **The MOS capacitor is made of three layers:** A metal contact (gate), an insulating oxide layer (typically silicon dioxide), and a semiconductor substrate.
- In the case of an n-type MOS capacitor, the semiconductor is doped with donor atoms that provide excess electrons (negative charge carriers).

**Operation:**

- **Flatband Condition:** When there is no voltage applied to the gate, the energy bands of the semiconductor are flat, and there is no bending of the energy levels. The system is neutral, and no excess charge accumulates at the interface.

- **Accumulation:** When a negative voltage is applied to the gate, it repels the electrons in the n-type semiconductor towards the interface, creating a high concentration of electrons near the oxide layer. This region is called the accumulation region.

- **Depletion:** When a positive voltage is applied, it attracts the electrons away from the surface of the semiconductor, leaving behind positively charged immobile donor ions. This creates a depletion region where the semiconductor is depleted of free carriers (electrons).

- **Inversion:** With a larger positive voltage, the surface potential becomes large enough to attract minority carriers (holes) to the surface. In the case of an n-type semiconductor, these holes form an inversion layer near the oxide-semiconductor interface. This inversion layer behaves like a p-type material, enabling current conduction similar to a p-type region.

**p-type Material is Usually Preferred due to following reason**

In MOSFETs and related applications, p-type material is often used for the substrate because the majority charge carriers are holes, and it allows for n-channel MOSFETs, which have higher electron mobility compared to p-channel MOSFETs. This results in faster switching speeds and better overall performance in digital circuits.

## **Question No.11**
## **Solution**

We have two sources of photons observed in the wavelength band from 400 nm to 600 nm. The photon distributions are:

Source A: $( n_A(\lambda) = A \lambda^3 )$

Source B: $( n_B(\lambda) = B \lambda^{-2} )$

Assuming the detector's quantum efficiency (QE) is not a function of wavelength, and the sources generate photoelectrons at the same rate, we need to compute their energy brightness ratio.


1. **Photon Energy Calculation**

   The energy \( E \) of a photon is given by:
   $$
   [
   E = \frac{hc}{\lambda}
   ]
   $$
   where:

      ( h \) is Planck's constant $(\(6.626 \times 10^{-34} \text{ Js}))$

      ( c \) is the speed of light $(\(3 \times 10^8 \text{ m/s}))$

2. **Total Energy Calculation for Each Source**

   The total energy $( E_{\text{total}} )$ is calculated by integrating the product of the photon flux and photon energy over the wavelength range:

   For Source A:
   $$
   [
   E_{\text{total, A}} = \int_{\lambda_{\text{min}}}^{\lambda_{\text{max}}} n_A(\lambda) \cdot E \, d\lambda
   ]
   $$

   $$
   [
   E_{\text{total, A}} = \int_{400 \text{ nm}}^{600 \text{ nm}} A \lambda^3 \cdot \frac{hc}{\lambda} \, d\lambda
   ]
   $$

   $$
   [
   E_{\text{total, A}} = A \cdot hc \int_{400 \text{ nm}}^{600 \text{ nm}} \lambda^2 \, d\lambda
   ]
   $$

   $$
   [
   E_{\text{total, A}} = A \cdot hc \left[ \frac{\lambda^3}{3} \right]_{400 \text{ nm}}^{600 \text{ nm}}
   ]
   $$

   $$
   [
   E_{\text{total, A}} = A \cdot hc \left( \frac{(600 \text{ nm})^3 - (400 \text{ nm})^3}{3} \right)
   ]
   $$

   For Source B:
   $$
   [
   E_{\text{total, B}} = \int_{\lambda_{\text{min}}}^{\lambda_{\text{max}}} n_B(\lambda) \cdot E \, d\lambda
   ]
   $$

   $$
   [
   E_{\text{total, B}} = \int_{400 \text{ nm}}^{600 \text{ nm}} B \lambda^{-2} \cdot \frac{hc}{\lambda} \, d\lambda
   ]
   $$

   $$
   [
   E_{\text{total, B}} = B \cdot hc \int_{400 \text{ nm}}^{600 \text{ nm}} \lambda^{-3} \, d\lambda
   ]
   $$

   $$
   [
   E_{\text{total, B}} = B \cdot hc \left[ -\frac{1}{2\lambda^2} \right]_{400 \text{ nm}}^{600 \text{ nm}}
   ]
   $$

   $$
   [
   E_{\text{total, B}} = B \cdot hc \left( -\frac{1}{2 (600 \text{ nm})^2} + \frac{1}{2 (400 \text{ nm})^2} \right)
   ]
   $$

3. **Energy Brightness Ratio**

   The energy brightness ratio \( R \) is given by:

   $$
   [
   R = \frac{E_{\text{total, A}}}{E_{\text{total, B}}}
   ]
   $$

   Substituting the integrals:

   $$
   [
   R = \frac{A \cdot hc \left( \frac{(600 \text{ nm})^3 - (400 \text{ nm})^3}{3} \right)}{B \cdot hc \left( -\frac{1}{2 (600 \text{ nm})^2} + \frac{1}{2 (400 \text{ nm})^2} \right)}
   ]
   $$

   $$
   [
   R = \frac{A \left( \frac{(600)^3 - (400)^3}{3} \right)}{B \left( -\frac{1}{2 (600)^2} + \frac{1}{2 (400)^2} \right)}
   ]
   $$

   $$

   [
   R = \frac{A \left( \frac{152000000}{3} \right)}{B \left( -\frac{1}{720000} + \frac{1}{320000} \right)}
   ]
   $$

   $$
   [
   R = \frac{A \cdot 50666666.67}{B \cdot \frac{1}{480000}}

## **Question No.12**
## **Solution**

#### **Part (a): Angular Resolution**

The diffraction-limited angular resolution $( \theta_D )$ is given by:

$$
[
\theta_D = \frac{1.22 \lambda}{D}
]
$$

Where:

- $( \lambda = 500 \, \text{nm} = 500 \times 10^{-9} \, \text{m} )$ for visible light considered,
- $( D = 0.7 \, \text{m} )$.

So,

**The angular resolution is:**

$$
[
\theta_D = \frac{1.22 \times 500 \times 10^{-9}}{0.7} \approx 8.71 \times 10^{-7} \, \text{radians} = 0.179 \, \text{arcsec}
]
$$

Since the diffraction-limited resolution is smaller than typical ground-based seeing (1 arcsec), the imaging is **seeing-limited**.


#### **Part (b): Flux of the Sky**

The sky brightness is given as $( \mu_{\text{bg}} = 21 \, \text{mag arcsec}^{-2} ).$

The flux $( F_{\text{sky}} )$ from a source with magnitude \( m \) is given by:

$$
[
F_{\text{sky}} = F_0 \times 10^{-0.4 \mu_{\text{bg}}}
]
$$

Where $( F_0 )$ is the reference flux, typically $( 3631 \, \text{Jy} )$ in the visible band.

Thus:

$$
[
F_{\text{sky}} = 3631 \times 10^{-0.4 \times 21}
]
$$

$$
[
F_{\text{sky}} \approx 3631 \times 10^{-8.4} = 3631 \times 3.981 \times 10^{-9} = 1.445 \times 10^{-5} \, \text{Jy/arcsec}^2
]
$$


#### **Part (c): Photon Rate from the Sky in the Star's Footprint**

The photon rate $( R_{\text{sky}} )$ can be calculated as:

$$
[
R_{\text{sky}} = F_{\text{sky}} \times A_{\text{footprint}}
]
$$

We calculate the footprint area:

$$
[
A_{\text{footprint}} = N_{\text{pix}} \times (\text{pixel scale})^2
]
$$

Where $( N_{\text{pix}} = 3.3 )$, and the pixel scale is calculated as:
$$
[
\text{Pixel scale} = \frac{\text{Pixel size}}{\text{Focal length}} = \frac{4 \times 10^{-6}}{3.5} \approx 1.14 \times 10^{-6} \, \text{radians} \approx 0.236 \, \text{arcsec/pixel}
]
$$

Thus,

**The footprint area is:**

$$
[
A_{\text{footprint}} \approx 3.3 \times (0.236)^2 = 3.3 \times 0.0557 = 0.184 \, \text{arcsec}^2
]
$$

**Finally, the photon rate:**
$$
[
R_{\text{sky}} = 1.445 \times 10^{-5} \times 0.184 = 2.66 \times 10^{-6} \, \text{Jy}
]
$$

#### **Part (d): Exposure Time for a V = 14.5 Star**

The ratio of fluxes for the V = 16 and V = 14.5 stars is:
$$
[
\frac{F_{14.5}}{F_{16}} = 10^{-0.4 (14.5 - 16)} = 10^{0.6} = 1.585
]
$$

The exposure time required for the V = 14.5 star is:
$$
[
t_{14.5} = \frac{t_{16}}{1.585} = \frac{30}{1.585} \approx 18.93 \, \text{seconds}
]
$$

#### **Part (e): Exposure Time for SNR = 200**

The exposure time needed to achieve an SNR of 200 can be calculated using:
$$
[
\frac{SNR_2}{SNR_1} = \sqrt{\frac{t_2}{t_1}}
]
$$

For $( SNR_2 = 200 ) and ( SNR_1 = 80 ):$
$$
[
\frac{200}{80} = \sqrt{\frac{t_2}{30}} \Rightarrow \left(\frac{200}{80}\right)^2 = \frac{t_2}{30}

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Constants
C_tilde = 1  # C̃ = 1
initial_conditions = [
    (1, 0.5, np.pi/2, 1),   # (r̃(0), r̃'(0), θ(0), θ'(0)) for case (a)
    (1, 1.0, np.pi/2, 1),   # case (b)
    (1, 1.03, np.pi/2, 1)   # case (c)
]

# Define the system of differential equations
def system(tau, y):
    r_tilde, r_tilde_dot, theta, theta_dot = y
    r_tilde_ddot = r_tilde * theta_dot**2 - C_tilde / r_tilde**2
    theta_ddot = -2 * r_tilde_dot * theta_dot / r_tilde
    return [r_tilde_dot, r_tilde_ddot, theta_dot, theta_ddot]

# Time span for integration
t_span = (0, 10)
num_points = 50
t_eval = np.linspace(t_span[0], t_span[1], num_points)

# Integrate for each case and plot the trajectory in the x-y plane
plt.figure(figsize=(10, 6))

for case in initial_conditions:
    sol = solve_ivp(system, t_span, case, t_eval=t_eval, method='RK45')

    r_tilde_vals = sol.y[0]
    theta_vals = sol.y[2]

    # Convert (r̃, θ) to (x, y)
    x_vals = r_tilde_vals * np.cos(theta_vals)
    y_vals = r_tilde_vals * np.sin(theta_vals)

    plt.plot(x_vals, y_vals, label=f'r̃(0)={case[0]}, r̃\'(0)={case[1]}')

plt.title('Conic Sections Trajectories')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()

# Function to compute Ẽ and l̃
def compute_energy_and_momentum(r_tilde, r_tilde_dot, theta_dot):
    E_tilde = 0.5 * (r_tilde_dot**2 + (r_tilde * theta_dot)**2) - C_tilde / r_tilde
    l_tilde = r_tilde**2 * theta_dot
    return E_tilde, l_tilde

# Compute energy and angular momentum for each case
for idx, case in enumerate(initial_conditions):
    E_tilde, l_tilde = compute_energy_and_momentum(case[0], case[1], case[3])
    print(f'Case {chr(97+idx)}: Ẽ = {E_tilde:.4f}, l̃ = {l_tilde:.4f}')
```

```python
# Compute eccentricity based on energy and angular momentum
def compute_eccentricity(E_tilde, l_tilde):
    return np.sqrt(1 + 2 * E_tilde * l_tilde**2 / C_tilde**2)

# Compute and print eccentricity for each case
for idx, case in enumerate(initial_conditions):
    E_tilde, l_tilde = compute_energy_and_momentum(case[0], case[1], case[3])
    eccentricity = compute_eccentricity(E_tilde, l_tilde)
    print(f'Case {chr(97+idx)}: Eccentricity = {eccentricity:.4f}')
```
