#### **Problem Statement**

Consider a rod of height \( L \) and mass \( m \). The lower end of the rod is at a distance \( R \) from the center of a spherical body with mass \( M \). We are asked to:

**(a)** Derive the expression for the tidal force experienced by the rod, assuming $( L \ll R ).$

**(b)** Calculate the tidal force if:
   - $( M = 1.3 \, \text{solar masses} = 1.3 \times 1.989 \times 10^{30} \, \text{kg} )$
   - $( R = 10 \, \text{km} = 10^4 \, \text{m} )$
   - The rod weighs $( m = 10 \, \text{kg} )$ and has a length $( L = 1 \, \text{m} ).$

**(c)** Determine the distance from the neutron star at which the rod would break, given a tensile strength of $( F_{\text{tensile}} = 10^4 \, \text{N} ).$



#### **Part (a): Tidal Force Derivation**

The gravitational force at the bottom of the rod, distance \( R \) from \( M \):
$$
[
F_{\text{bottom}} = \frac{G M m}{R^2}
]
$$

The gravitational force at the top of the rod, at distance \( R + L \) from \( M \):
$$
[
F_{\text{top}} = \frac{G M m}{(R + L)^2}
]
$$

The tidal force experienced by the rod is the difference:
$$
[
F_{\text{tidal}} = F_{\text{bottom}} - F_{\text{top}}
]
$$

Since $( L \ll R )$, we can use the binomial expansion $( (R + L)^{-2} \approx \frac{1}{R^2} - \frac{2L}{R^3} ):$
$$
[
F_{\text{tidal}} \approx \frac{G M m}{R^2} - \left( \frac{G M m}{R^2} - \frac{2 G M m L}{R^3} \right)
]
$$

Simplifying, we get:
$$
[
F_{\text{tidal}} \approx \frac{2 G M m L}{R^3}
]
$$

#### **Part (b): Tidal Force Calculation**

Using the values:
$$
[
G = 6.674 \times 10^{-11} \, \text{Nm}^2/\text{kg}^2, \quad M = 1.3 \times 1.989 \times 10^{30} \, \text{kg}, \quad m = 10 \, \text{kg}, \quad L = 1 \, \text{m}, \quad R = 10^4 \, \text{m}
]
$$

Substituting into the tidal force formula:
$$
[
F_{\text{tidal}} \approx \frac{2 \times (6.674 \times 10^{-11}) \times (1.3 \times 1.989 \times 10^{30}) \times 10 \times 1}{(10^4)^3}
]
$$

Calculating the result:
$$
[
F_{\text{tidal}} \approx 3.45 \times 10^9 \, \text{N} \approx 3451392360.0  N
]
$$


#### **Part (c): Distance for Rod Breakup**

We need to find the distance \( R \) where $( F_{\text{tidal}} )$ equals the rod's tensile strength $( F_{\text{tensile}} = 10^4 \, \text{N} ).$

Set $( F_{\text{tidal}} = F_{\text{tensile}} )$:

$$
[
\frac{2 G M m L}{R^3} = F_{\text{tensile}}
]
$$

Solving for \( R \):
$$
[
R^3 = \frac{2 G M m L}{F_{\text{tensile}}}
]
$$
$$
[
R = \left( \frac{2 \times (6.674 \times 10^{-11}) \times (1.3 \times 1.989 \times 10^{30}) \times 10 \times 1}{10^4} \right)^{\frac{1}{3}}
]
$$

Calculating the result:
$$
[
R \approx 7.01452247611302  \times 10^5 \, \text{m} ≈ 701452.248 m
]
$$

- **Tidal Force Formula**: $( F_{\text{tidal}} \approx \frac{2 G M m L}{R^3} )$
- **Tidal Force on Neutron Star Surface**: $( F_{\text{tidal}} \approx 3.45 \times 10^9 \, \text{N} \approx 3451392360.0  N )$

#### **Part (a) - Axisymmetric System and Simplification of $(J_n)$**

**Axisymmetry:**

An axisymmetric system is one that is invariant under rotations around a specific axis, typically the z-axis. In this case, axisymmetry means that the mass distribution of the planet does not depend on the azimuthal angle $(\phi)$, but only on the polar angle $(\theta)$. Thus, the system is symmetric about the z-axis, and the surface of the planet and its mass distribution are symmetric about this axis.

#### **Problem Recap**

We are given the external gravitational potential $( V_{\text{ext}}(d, \theta) )$ for a planet with radially varying mass density:
$$
[
V_{\text{ext}}(d, \theta) = -\frac{GM}{d} \left[ 1 - \sum_{n=2}^{\infty} J_n \left( \frac{R_p}{d} \right)^n P_n(\cos \theta) \right]
]
$$

where the coefficients $( J_n )$ are defined as:
$$
[
J_n = \frac{1}{M_p R_p^n} \int_0^R \int_0^{2\pi} \int_0^{\pi} r^n \rho(r, \theta) P_n(\cos \theta) r^2 \sin \theta \, d\theta \, d\phi \, dr
]
$$

In this expression, $( \rho(r, \theta) )$ is the mass density, which could depend on both $( r )$ and $( \theta )$, and $( M_p )$ is the total mass of the planet.

#### **Axisymmetric Assumption**

Assuming axisymmetry, the density $( \rho(r, \theta) )$ depends only on $( r )$, i.e., $( \rho(r, \theta) = \rho(r) )$. This allows us to perform the $( \phi )$-integration directly:
$$
[
J_n = \frac{2\pi}{M_p R_p^n} \int_0^R \int_0^{\pi} r^{n+2} \rho(r) P_n(\cos \theta) \sin \theta \, d\theta \, dr
]
$$

#### **Step 1: Integration Over $( \theta )$**
$$
[
I_\theta = \int_0^{\pi} P_n(\cos \theta) \sin \theta \, d\theta
]
$$

This integral can be evaluated using the properties of Legendre polynomials. For $( n \geq 2 )$, the result is:
$$
[
I_\theta = \int_0^{\pi} P_n(\cos \theta) \sin \theta \, d\theta = \frac{2}{2n+1}
]
$$

#### **Step 2: Substitute Back into $( J_n )$**

Substitute the result of the $( \theta )$-integration into the expression for $( J_n )$:
$$
[
J_n = \frac{2\pi}{M_p R_p^n} \cdot \frac{2}{2n+1} \int_0^R r^{n+2} \rho(r) \, dr
]
$$

Simplifying, we have:
$$
[
J_n = \frac{4\pi}{(2n+1) M_p R_p^n} \int_0^R r^{n+2} \rho(r) \, dr
]
$$

#### **Step 3: Radial Integral**

Now, we evaluate the radial integral:
$$
[
\int_0^R r^{n+2} \rho(r) \, dr
]
$$

If $( \rho(r) )$ is constant (uniform density), say $( \rho(r) = \rho_0 )$, then:
$$
[
\int_0^R r^{n+2} \rho_0 \, dr = \rho_0 \cdot \frac{R^{n+3}}{n+3}
]
$$

#### **Step 4: Substitute Constant Density Result**

With constant density, $( J_n )$ becomes:
$$
[
J_n = \frac{4\pi \rho_0}{(2n+1) M_p R_p^n} \cdot \frac{R^{n+3}}{n+3}
]
$$

Simplifying further:
$$
[
J_n = \frac{4\pi \rho_0}{(2n+1)(n+3) M_p} \cdot R_p^3
]
$$

#### **Final Answer**

In summary, the general expression for $( J_n )$ in an axisymmetric system is:
$$
[
J_n = \frac{4\pi}{(2n+1) M_p R_p^n} \int_0^R r^{n+2} \rho(r) \, dr
]
$$

For a constant density $( \rho_0 )$:
$$
[
J_n = \frac{4\pi \rho_0}{(2n+1)(n+3) M_p} \cdot R_p^3
]
$$

#### **Solution for $( J_n )$ Using $( \mu = \cos \theta )$**

We start with the expression for $( J_n )$ in an axisymmetric case:
$$
[
J_n = \frac{2\pi}{M_p R_p^n} \int_0^R \int_0^{\pi} r^{n+2} \rho(r) P_n(\cos \theta) \sin \theta \, d\theta \, dr
]
$$

#### **Step 1: Substitute $( \mu = \cos \theta )$**

To simplify the integral, we substitute $( \mu = \cos \theta )$, where $( \mu )$ is the new variable of integration.

- Under this substitution, when $( \theta = 0 )$, we have $( \mu = 1 )$, and when $( \theta = \pi )$, we have $( \mu = -1 )$.

- Therefore, $( \mu )$ varies from $( 1 )$ to $( -1 )$ as $( \theta )$ goes from $( 0 )$ to $( \pi )$.

#### **Step 2: Change in Differential**

Since $( \mu = \cos \theta )$, we can differentiate both sides with respect to $( \theta )$:
$$
[
d\mu = -\sin \theta \, d\theta
]
$$

This implies that:
$$
[
\sin \theta \, d\theta = -d\mu
]
$$

#### **Step 3: Rewriting the Integral in Terms of $( \mu )$**

With this substitution, the integral over $( \theta )$ from $( 0 )$ to $( \pi )$ becomes an integral over $( \mu )$ from $( 1 )$ to $( -1 )$.
Thus, the expression for $( J_n )$ becomes:
$$
[
J_n = \frac{2\pi}{M_p R_p^n} \int_0^R r^{n+2} \rho(r) \left( \int_{1}^{-1} P_n(\mu) \cdot (-d\mu) \right) dr
]
$$
Simplifying the sign, we obtain:
$$
[
J_n = \frac{2\pi}{M_p R_p^n} \int_0^R r^{n+2} \rho(r) \left( \int_{-1}^{1} P_n(\mu) \, d\mu \right) dr
]
$$

#### **Step 4: Evaluate the Integral Over $( \mu )$**

Now, we focus on evaluating the integral over $( \mu )$:
$$
[
\int_{-1}^{1} P_n(\mu) \, d\mu
]
$$

This integral depends on the properties of Legendre polynomials $( P_n(\mu) )$.

#### **Key Property of Legendre Polynomials**

For $( n \geq 2 )$, it is known that:
$$
[
\int_{-1}^{1} P_n(\mu) \, d\mu = 0
]
$$

This result follows from the orthogonality of Legendre polynomials over the interval $( [-1, 1] )$, where only the zeroth polynomial $( P_0(\mu) = 1 )$ contributes a non-zero integral over this interval.

#### **Step 5: Conclusion for $( J_n )$**

Since $( \int_{-1}^{1} P_n(\mu) \, d\mu = 0 )$ for all $( n \geq 2 )$, we find that:
$$
[
J_n = 0 \quad \text{for all} \quad n \geq 2
]
$$

#### **Final Result: Simplified Gravitational Potential**

Given that $( J_n = 0 )$ for $( n \geq 2 )$, the external gravitational potential $( V_{\text{ext}}(d, \theta) )$ simplifies significantly:
$$
[
V_{\text{ext}}(d, \theta) = -\frac{GM}{d}
]
$$

since the summation term in the potential expression does not contribute for $( n \geq 2 )$ in an axisymmetric system.

#### **Part (a) Self Garvity**

**(i) Potential due to self-gravity if the planet is spherical:**

For a spherical planet, the gravitational potential at a point ( P ) on the surface is given by:
$$
[
V_{\text{selfg}}(r) = - \frac{GM}{r}
]
$$

where:
- ( G \) is the gravitational constant,
- ( M \) is the mass of the planet,
- ( r \) is the radial distance from the center of the planet to the point \( P \).

Since the planet is spherical, $( r = R_p )$, the radius of the planet, so the gravitational potential at point $( P )$ is:
$$
[
V_{\text{selfg}}(R_p) = - \frac{GM}{R_p}
]
$$

**(ii) Potential due to self-gravity if the planet is an oblate spheroid:**

For an oblate spheroid, the radial distance varies with the angle $( \theta )$ as:
$$
[
r = R_p(1 + \epsilon \sin^2 \theta)
]
$$

where $( \epsilon )$ is a small parameter that measures the oblatness of the planet. The gravitational potential at point ( P ) due to self-gravity in the case of an oblate spheroid can be expanded as:
$$
[
V_{\text{selfg}}(r, \theta) = - \frac{GM}{r} \left( 1 - \frac{3}{2} \epsilon \sin^2 \theta \right)
]
$$

Substituting $( r = R_p(1 + \epsilon \sin^2 \theta) )$, we get the full potential:

$$
[
V_{\text{selfg}}(R_p, \theta) = - \frac{GM}{R_p(1 + \epsilon \sin^2 \theta)} \left( 1 - \frac{3}{2} \epsilon \sin^2 \theta \right)
]
$$

#### **Part (c) Tidal Potential**


**(i) Centrifugal Force and Potential:**

The centrifugal force per unit mass at a point ( P ) on the surface of the rotating planet is given by:

$$
[
\mathbf{F}_{\text{cf}} = \Omega \times (\Omega \times \mathbf{r})
]
$$

where $( \Omega )$ is the angular velocity vector of the planet, and $( \mathbf{r} )$ is the position vector of the point ( P ).

Since $( \Omega )$ is along the ( z )-axis, we have:

$$
[
\Omega = \Omega \hat{z}
]
$$

The position vector $( \mathbf{r} )$ in spherical coordinates is:
$$
[
\mathbf{r} = r \hat{r}
]
$$

Now, calculate the cross product $( \Omega \times \mathbf{r} )$:

$$
[
\Omega \times \mathbf{r} = \Omega R_p \sin \theta \, \hat{\phi}
]
$$

Next, calculate the cross product again:
$$
[
\Omega \times (\Omega \times \mathbf{r}) = \Omega^2 R_p \sin^2 \theta \, \hat{r}
]
$$

This gives the centrifugal force per unit mass directed radially outward from the center of the planet, with magnitude $( \Omega^2 R_p \sin^2 \theta )$.

The centrifugal potential is related to the centrifugal force by:
$$
[
\mathbf{F}_{\text{cf}} = - \nabla_r V_{\text{cf}}
]
$$

Assume the centrifugal potential to be:

$$
[
V_{\text{cf}}(r, \theta) = -\frac{1}{2} \Omega^2 r^2 \sin^2 \theta
]
$$
Now, compute the gradient of $( V_{\text{cf}} )$:
$$
[
\nabla_r V_{\text{cf}} = -\Omega^2 r \sin^2 \theta \hat{r} - \Omega^2 r \sin \theta \cos \theta \hat{\theta}
]
$$

Thus, the centrifugal force is:

$$
[
\mathbf{F}_{\text{cf}} = \nabla_r V_{\text{cf}}
]
$$

This confirms that the centrifugal force can be written as the gradient of the centrifugal potential.

#### **(a)**
The differential equation for a forced-damped harmonic oscillator is:

$$
[
\ddot{x} + \beta \dot{x} + \omega_0^2 x = F e^{i \omega t}
]$$

For part (a), we are considering the case where $( F = 0 )$, so the equation reduces to the homogeneous form:
$$
[
\ddot{x} + \beta \dot{x} + \omega_0^2 x = 0
]
$$

#### **Step 1: Form of the Solution**

We assume a solution of the form:

$$
[
x(t) = e^{\lambda t}
]
$$

where $( \lambda )$ is a constant to be determined. Substituting $( x(t) = e^{\lambda t} )$ into the differential equation gives:

$$
[
\lambda^2 e^{\lambda t} + \beta \lambda e^{\lambda t} + \omega_0^2 e^{\lambda t} = 0
]
$$

Dividing through by $( e^{\lambda t} )$ (which is non-zero), we obtain the characteristic equation:

$$
[
\lambda^2 + \beta \lambda + \omega_0^2 = 0
]
$$

#### **Step 2: Solve the Characteristic Equation**

The characteristic equation is a quadratic equation in $( \lambda )$:

$$
[
\lambda^2 + \beta \lambda + \omega_0^2 = 0
]
$$

Using the quadratic formula $( \lambda = \frac{-b \pm \sqrt{b^2 - 4ac}}{2a} )$, where $( a = 1 )$, $( b = \beta )$, and $( c = \omega_0^2 )$, we get:

$$
[
\lambda = \frac{-\beta \pm \sqrt{\beta^2 - 4 \omega_0^2}}{2}
]
$$

#### **Step 3: Analyze the Roots of the Characteristic Equation**

The nature of the roots $( \lambda )$ depends on the discriminant $( \beta^2 - 4 \omega_0^2 )$:

1. **Overdamped Case** $(( \beta^2 > 4 \omega_0^2 )$): The roots are real and distinct.
$$
   [
   \lambda_1 = \frac{-\beta + \sqrt{\beta^2 - 4 \omega_0^2}}{2}, \quad \lambda_2 = \frac{-\beta - \sqrt{\beta^2 - 4 \omega_0^2}}{2}
   ]
$$
   The solution in this case is:
$$
   [
   x(t) = C_1 e^{\lambda_1 t} + C_2 e^{\lambda_2 t}
   ]
$$

2. **Critically Damped Case** $(( \beta^2 = 4 \omega_0^2 ))$: The roots are real and equal.
$$
   [
   \lambda = -\frac{\beta}{2}
   ]
$$
   The solution in this case is:
$$
   [
   x(t) = (C_1 + C_2 t) e^{-\frac{\beta}{2} t}
   ]
$$

3. **Underdamped Case** $(( \beta^2 < 4 \omega_0^2 ))$: The roots are complex conjugates.
$$
   [
   \lambda = -\frac{\beta}{2} \pm i \omega_d
   ]
$$
   where $( \omega_d = \sqrt{\omega_0^2 - \frac{\beta^2}{4}} )$ is the damped natural frequency. The solution in this case is:
$$
   [
   x(t) = e^{-\frac{\beta}{2} t} \left( C_1 \cos(\omega_d t) + C_2 \sin(\omega_d t) \right)
   ]
$$

#### **Summary of the Homogeneous Solution**

The solution to the homogeneous equation depends on the damping parameter $( \beta )$ relative to $( \omega_0 )$:

1. **Overdamped** $(( \beta^2 > 4 \omega_0^2 ))$:
$$
   [
   x(t) = C_1 e^{\lambda_1 t} + C_2 e^{\lambda_2 t}
   ]
$$

2. **Critically Damped** $(( \beta^2 = 4 \omega_0^2 ))$:
$$
   [
   x(t) = (C_1 + C_2 t) e^{-\frac{\beta}{2} t}
   ]
$$

3. **Underdamped** $(( \beta^2 < 4 \omega_0^2 ))$:
$$
   [
   x(t) = e^{-\frac{\beta}{2} t} \left( C_1 \cos(\omega_d t) + C_2 \sin(\omega_d t) \right)
   ]
$$
   where $( \omega_d = \sqrt{\omega_0^2 - \frac{\beta^2}{4}} )$ is the damped natural frequency.


#### **(b) Long-Term Behavior (t → ∞)**

We start with the forced-damped harmonic oscillator equation:
$$
[
\ddot{x} + \beta \dot{x} + \omega_0^2 x = F e^{i \omega t}
]
$$

**Step 1: General Solution as the Sum of Homogeneous and Particular Solutions**

The total solution $( x(t) )$ is the sum of the homogeneous solution $( x_h(t) )$ and the particular solution $( x_p(t) )$:

$$
[
x(t) = x_h(t) + x_p(t)
]
$$

As $( t \to \infty )$, the homogeneous solution $( x_h(t) )$, which contains transient terms, dies out. Thus, we only need to focus on finding the particular solution $( x_p(t) )$ for the long-term behavior.

**Step 2: Assume a Particular Solution Form**

To solve for $( x_p(t) )$, assume a solution of the form:

$$
[
x_p(t) = A e^{i(\omega t - \delta)}
]
$$

where \( A \) is the amplitude and $( \delta )$ is the phase shift. We will determine \( A \) and $( \delta )$ by substituting $( x_p(t) )$ into the original differential equation.

**Step 3: Substitute $( x_p(t) )$ into the Differential Equation**

Compute $( \dot{x}_p(t) )$ and $( \ddot{x}_p(t) ):$
$$
[
\dot{x}_p(t) = i \omega A e^{i(\omega t - \delta)}
]
$$
$$
[
\ddot{x}_p(t) = -\omega^2 A e^{i(\omega t - \delta)}
]
$$

Substitute these expressions into the differential equation:

$$
[
-\omega^2 A e^{i(\omega t - \delta)} + \beta (i \omega A e^{i(\omega t - \delta)}) + \omega_0^2 A e^{i(\omega t - \delta)} = F e^{i \omega t}
]
$$

Factor out $( A e^{i(\omega t - \delta)} ):$

$$
[
A \left( -\omega^2 + i \beta \omega + \omega_0^2 \right) e^{i(\omega t - \delta)} = F e^{i \omega t}
]
$$

**Step 4: Equate Terms and Solve for ( A ) and $( \delta )$**

For this equation to hold, the exponential terms on both sides must match. Thus, we set:

$$
[
A \left( -\omega^2 + i \beta \omega + \omega_0^2 \right) = F
]
$$

This is a complex equation. We can rewrite it by separating real and imaginary parts:
$$
[
A (\omega_0^2 - \omega^2) + i A \beta \omega = F
]
$$

**Step 5: Amplitude ( A \) and Phase Shift $( \delta )$**

Rewrite \( A \) in terms of \( F \), $( \omega )$, $( \omega_0 )$, and $( \beta )$:

$$
[
A = \frac{F}{\sqrt{(\omega_0^2 - \omega^2)^2 + (\beta \omega)^2}}
]
$$

The phase shift $( \delta )$ is given by:

$$
[
\delta = \tan^{-1} \left( \frac{\beta \omega}{\omega_0^2 - \omega^2} \right)
]
$$

**Final Solution**

Thus, the solution in the late-time limit (after transient solutions have died down) is:
$$
[
x(t) = A e^{i(\omega t - \delta)}
]
$$

where
$$
[
A = \frac{F}{\sqrt{(\omega_0^2 - \omega^2)^2 + (\beta \omega)^2}}
]
$$
and
$$
[
\delta = \tan^{-1} \left( \frac{\beta \omega}{\omega_0^2 - \omega^2} \right)
]
$$
