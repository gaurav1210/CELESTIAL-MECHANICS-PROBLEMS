## **Question No.1**
### **Solution**


We want to prove that in polar coordinates:
$$
[
\dot{\mathbf{r}} = \dot{r} \hat{r} + r \dot{\theta} \hat{\theta}
]
$$

and
$$
[
\ddot{\mathbf{r}} = (\ddot{r} - r \dot{\theta}^2) \hat{r} + (r \ddot{\theta} + 2 \dot{r} \dot{\theta}) \hat{\theta}
]
$$

#### **Step 1: Position Vector in Polar Coordinates**

The position vector $( \mathbf{r} )$ of a point \( P \) in polar coordinates is:
$$
[
\mathbf{r} = r \hat{r}
]
$$

where \( r \) is the radial distance, and $( \hat{r} )$ is the unit vector in the radial direction. The unit vector in Cartesian coordinates is given by:

$$
[
\hat{r} = \cos \theta \, \hat{i} + \sin \theta \, \hat{j}
]
$$

The unit vector $( \hat{\theta} )$, perpendicular to $( \hat{r} )$, is:
$$
[
\hat{\theta} = -\sin \theta \, \hat{i} + \cos \theta \, \hat{j}
]
$$

Thus, the position vector becomes:
$$
[
\mathbf{r} = r (\cos \theta \, \hat{i} + \sin \theta \, \hat{j})
]
$$
#### **Step 2: Velocity $( \dot{\mathbf{r}} )$**

To find the velocity, we take the time derivative of $( \mathbf{r} )$:
$$
[
\dot{\mathbf{r}} = \frac{d}{dt} (r \hat{r})
]
$$

Using the product rule, we get:
$$
[
\dot{\mathbf{r}} = \dot{r} \hat{r} + r \frac{d}{dt} \hat{r}
]
$$

#### **Step 3: Derivative of $( \hat{r} )$**

Since $( \hat{r} = \cos \theta \, \hat{i} + \sin \theta \, \hat{j} )$,

we apply the chain rule to differentiate:

$$[
\frac{d}{dt} \hat{r} = -\sin \theta \, \dot{\theta} \, \hat{i} + \cos \theta \, \dot{\theta} \, \hat{j}
]
$$

This simplifies to:

$$
[
\frac{d}{dt} \hat{r} = \dot{\theta} \hat{\theta}
]
$$

Thus:

$$
[
\dot{\mathbf{r}} = \dot{r} \hat{r} + r \dot{\theta} \hat{\theta}
]
$$

#### **Step 4: Acceleration $( \ddot{\mathbf{r}} )$**

The acceleration is the time derivative of the velocity:
$$
[
\ddot{\mathbf{r}} = \frac{d}{dt} (\dot{r} \hat{r} + r \dot{\theta} \hat{\theta})
]
$$
Using the product rule again:
$$
[
\ddot{\mathbf{r}} = \frac{d}{dt} (\dot{r} \hat{r}) + \frac{d}{dt} (r \dot{\theta} \hat{\theta})
]
$$

First, we differentiate $( \dot{r} \hat{r} )$:

$$
[
\frac{d}{dt} (\dot{r} \hat{r}) = \ddot{r} \hat{r} + \dot{r} \dot{\theta} \hat{\theta}
]
$$

Next, we differentiate $( r \dot{\theta} \hat{\theta} )$:
$$
[
\frac{d}{dt} (r \dot{\theta} \hat{\theta}) = (\dot{r} \dot{\theta} + r \ddot{\theta}) \hat{\theta} + r \dot{\theta} (-\dot{\theta} \hat{r})
]
$$

Simplifying:
$$
[
\frac{d}{dt} (r \dot{\theta} \hat{\theta}) = (\dot{r} \dot{\theta} + r \ddot{\theta}) \hat{\theta} - r \dot{\theta}^2 \hat{r}
]
$$

#### **Step 5: Final Expression for Acceleration**

Combining both terms, we get:

$$
[
\ddot{\mathbf{r}} = (\ddot{r} - r \dot{\theta}^2) \hat{r} + (r \ddot{\theta} + 2 \dot{r} \dot{\theta}) \hat{\theta}
]
$$

Thus,

**The velocity and acceleration in polar coordinates are:**

$$
[
\dot{\mathbf{r}} = \dot{r} \hat{r} + r \dot{\theta} \hat{\theta}
]
$$
$$
[
\ddot{\mathbf{r}} = (\ddot{r} - r \dot{\theta}^2) \hat{r} + (r \ddot{\theta} + 2 \dot{r} \dot{\theta}) \hat{\theta}
]
$$ **Proved**


## **Question No.2**
### **Solution**

 Prove that the total mechanical energy
 $$
[
E = \frac{1}{2} \mu \mathbf{v} \cdot \mathbf{v} + U(r)
]
$$
is a constant of motion for any central force in the two-body problem.

---

#### **Step-by-Step Mathematical Derivation:***

1. **Total Energy Expression**:  

   The total energy \( E \) is given by:
   $$
   [
   E = \frac{1}{2} \mu \mathbf{v} \cdot \mathbf{v} + U(r)
   ]
   $$

   where $( \mathbf{v} = \dot{\mathbf{r}} )$ is the velocity of the particle and $( \mathbf{r} )$ is its position vector.

2. **Time Derivative of Energy**:

   To show that \( E \) is conserved, we take the time derivative of the total energy \( E \):
   $$
   [
   \frac{dE}{dt} = \frac{d}{dt} \left( \frac{1}{2} \mu \mathbf{v} \cdot \mathbf{v} + U(r) \right)
   ]
   $$

   This breaks into two parts:
   $$
   [
   \frac{dE}{dt} = \mu \mathbf{v} \cdot \frac{d\mathbf{v}}{dt} + \frac{dU(r)}{dt}
   ]
   $$

3. **Time Derivative of Potential Energy**:

   The potential energy $( U(r) )$ depends only on the radial distance \( r \), so:
   $$
   [
   \frac{dU(r)}{dt} = \frac{dU}{dr} \cdot \frac{dr}{dt} = \frac{dU}{dr} \cdot \dot{r}
   ]
   $$

   But from the definition of the central force $( \mathbf{F}(r) )$, we know that:
   $$
   [
   \mathbf{F}(r) = -\frac{dU}{dr} \hat{r}
   ]
   $$
   Hence,
   $$
   [
   \frac{dU(r)}{dt} = - \mathbf{F}(r) \cdot \mathbf{v}
   ]
   $$
   where $( \mathbf{v} = \dot{\mathbf{r}} ).$

4. **Time Derivative of Kinetic Energy**:  

   The kinetic energy term is:
   $$
   [
   \mu \mathbf{v} \cdot \frac{d\mathbf{v}}{dt}
   ]
   $$
   Using Newton's second law, $( \mu \frac{d\mathbf{v}}{dt} = \mathbf{F}(r) )$, we substitute this into the above equation:
   $$
   [
   \mu \mathbf{v} \cdot \frac{d\mathbf{v}}{dt} = \mathbf{v} \cdot \mathbf{F}(r)
   ]
  $$
5. **Total Derivative of Energy**:

   Combining the two results, the total time derivative of energy is:
   $$
   [
   \frac{dE}{dt} = \mathbf{v} \cdot \mathbf{F}(r) - \mathbf{v} \cdot \mathbf{F}(r)
   ]
   $$
   This simplifies to:
   $$
   [
   \frac{dE}{dt} = 0
   ]
   $$
Since the time derivative of the total energy is zero, we conclude that the total energy \( E \) is constant:
$$
[
E = \text{constant}
]
$$

#### **Conclusion:**

We have mathematically shown that the total mechanical energy
$$
[
E = \frac{1}{2} \mu \mathbf{v} \cdot \mathbf{v} + U(r)
]
$$
is conserved for any central force $( \mathbf{F}(r) )$, as the time derivative of the total energy is zero.

#### **Step-by-Step Derivation of $( r(\theta) )$**

**1. Equations of Motion in Polar Coordinates:**

The motion of a particle in polar coordinates is described by:
$$
[
\ddot{r} = r \dot{\theta}^2 - \frac{F_r}{\mu} \tag{1}
]
$$
$$
[
\frac{d}{dt}(r^2 \dot{\theta}) = 0 \tag{2}
]
$$

where $( F_r )$ is the radial force, $( \mu )$ is the reduced mass, and $( \dot{\theta} )$ is the angular velocity. The angular momentum conservation gives:
$$
[
r^2 \dot{\theta} = h = \text{constant} \tag{3}
]
$$

**2. Solve for $( \dot{\theta} )$:**

From the conservation of angular momentum, we have:
$$
[
\dot{\theta} = \frac{h}{r^2} \tag{4}
]
$$

**3. Substitute $( \dot{\theta} )$** into the Radial Equation:

Substituting equation (4) into equation (1) gives us:
$$
[
\ddot{r} = \frac{h^2}{r^3} - \frac{F_r}{\mu} \tag{5}
]
$$

**4. Substitution of $( u = \frac{1}{r} )$:**

Now we substitute $( u = \frac{1}{r} )$, leading to:
$$
[
r = \frac{1}{u}, \quad \dot{r} = -\frac{1}{u^2} \frac{du}{dt}
]
$$
The second derivative of \( r \) with respect to time is:
$$
[
\ddot{r} = \frac{2}{u^3} \left( \frac{du}{dt} \right)^2 - \frac{1}{u^2} \frac{d^2u}{dt^2}
]
$$
Since $( \dot{\theta} = \frac{h}{r^2} = h u^2 )$, using the chain rule:
$$
[
\frac{du}{dt} = \frac{du}{d\theta} \cdot h u^2
]
$$

Thus, the second derivative of \( r \) becomes:
$$
[
\ddot{r} = h^2 u^2 \left( \frac{d^2u}{d\theta^2} + u \right) \tag{6}
]
$$

**5. Substitute Back into the Radial Equation:**

Substitute equation (6) into equation (5):
$$
[
h^2 u^2 \left( \frac{d^2u}{d\theta^2} + u \right) = \frac{h^2 u^3}{\mu} - F_r
]
$$

This simplifies to:
$$
[
\frac{d^2u}{d\theta^2} + u = -\frac{F_r}{h^2 u^2} \tag{7}
]
$$

**6. Solve the Differential Equation:**

Assume a central force $( F_r = -\frac{k}{r^2} = -k u^2 )$, we get:
$$
[
\frac{d^2u}{d\theta^2} + u = \frac{k}{h^2} \tag{8}
]
$$

The general solution is:
$$
[
u(\theta) = C \cos(\theta - \theta_0) + \frac{k}{h^2}
]
$$

**7. Substitute Back for ( r ):**

Since $( u = \frac{1}{r} )$, the final solution for \( r \) is:

$$
[
r(\theta) = \frac{1}{C \cos(\theta - \theta_0) + \frac{k}{h^2}}
]
$$

#### **Conclusion:**

We have derived $( r(\theta) = \frac{1}{C \cos(\theta - \theta_0) + \frac{k}{h^2}} )$, representing the general solution for the radial distance in polar coordinates.


## **Question No.4**

**Proof that \( E < 0 \) Results in Elliptical Orbits (Keplerâ€™s First Law)**

**Step 1: Energy and Force Relation**

The total energy \( E \) of a two-body system can be expressed as:
$$
[
E = \frac{1}{2} \mu v^2 - \frac{k}{r}
]
$$
where:
- $( \mu )$ is the reduced mass of the two-body system.
- \( v \) is the relative velocity between the two masses.
- \( k \) is a positive constant representing the strength of the central force (gravitational force, for example).
- \( r \) is the distance between the two masses.

**Step 2: Conditions for Elliptical Trajectories**

For an elliptical orbit, the total energy must be negative:
$$
[
E < 0
]
$$
This implies:
$$
[
\frac{1}{2} \mu v^2 < \frac{k}{r} \tag{1}
]
$$

This inequality means that the kinetic energy of the system is less than the potential energy due to the central force. Hence, the system is bound, leading to elliptical orbits.

**Step 3: Conservation of Angular Momentum**

The angular momentum \( L \) of the system is given by:
$$
[
L = \mu r^2 \dot{\theta}
]
$$

where $( \dot{\theta} )$ is the angular velocity. The conservation of angular momentum states that \( L \) is constant.

**Step 4: Relationship Between Energy, Potential, and Angular Momentum
From equations of motion in polar coordinates,**

We have:

$$
[
\dot{r} = \frac{dr}{dt}, \quad \dot{\theta} = \frac{d\theta}{dt}
]
$$

We can express the kinetic energy in terms of \( r \) and $( \theta ):$
$$
[
v^2 = \dot{r}^2 + (r \dot{\theta})^2
]
$$

Substituting $( \dot{\theta} = \frac{L}{\mu r^2} ):$
$$
[
v^2 = \dot{r}^2 + \left( \frac{L}{\mu r} \right)^2 \tag{2}
]
$$

**Step 5: Substituting into the Energy Equation**
Substituting equation (2) into the expression for energy (1):

$$
[
E = \frac{1}{2} \mu \left( \dot{r}^2 + \frac{L^2}{\mu^2 r^2} \right) - \frac{k}{r}
]
$$

This can be rearranged to:
$$
[
E = \frac{1}{2} \mu \dot{r}^2 + \frac{L^2}{2 \mu r^2} - \frac{k}{r}
]
$$

**Step 6: Use the Effective Potential**

Define the effective potential $( U_{\text{eff}}(r) ):$

$$
[
U_{\text{eff}}(r) = -\frac{k}{r} + \frac{L^2}{2 \mu r^2}
]
$$

Thus, we have:
$$
[
E = \frac{1}{2} \mu \dot{r}^2 + U_{\text{eff}}(r)
]
$$

**Step 7: Analyze the Potential Function**

The effective potential has a minimum, indicating a stable orbit. The condition $( E < 0 )$ means that the system is trapped in a potential well formed by $( U_{\text{eff}}(r) ).$

**Step 8: Deriving the Equation of the Orbit**
Using the equation for effective potential and conservation laws, we can derive the equation of motion in the form of a conic section:
$$
[
\frac{d^2\theta}{dr^2} + \frac{1}{r} \frac{d\theta}{dr} = -\frac{k}{L^2} r
]
$$

This results in a standard conic section equation, which describes an ellipse when $( E < 0 ).$

**Step 9: Conclusion**

The final expression shows that the trajectory of the radius vector $( \mathbf{r} )$ joining the two masses describes an ellipse, with the focus of the ellipse located at the origin (the center of mass of the system). This conclusion is consistent with Kepler's first law of planetary motion, stating that planets move in elliptical orbits with the sun at one focus.

**Summary of the Result**

Thus, when the energy $( E < 0 ):$
$$
[
\mathbf{r} \text{ traces an ellipse with the origin as a focus.}
]
$$
This completes the proof that negative total energy results in elliptical orbits for the two-body problem under central forces.

**Proof that the Velocity of the Reduced Mass is Maximum at Perihelion**

**Step 1: Define the System**

Consider two masses $( m_1 )$ and $( m_2 )$ in a central force field with a mutual distance \( r \) between them. The reduced mass $( \mu )$ is defined as:
$$
[
\mu = \frac{m_1 m_2}{m_1 + m_2}
]
$$
**Step 2: Total Energy of the System**

The total mechanical energy \( E \) of the two-body system can be expressed as:
$$
[
E = \frac{1}{2} \mu v^2 - \frac{k}{r}
]
$$
where ( k ) is a constant related to the gravitational force $(e.g., ( k = G m_1 m_2 ))$ and \( v \) is the relative velocity.

**Step 3: Conservation of Angular Momentum**

The angular momentum \( L \) of the system is given by:
$$
[
L = \mu r^2 \dot{\theta}
]
$$

where $( \dot{\theta} )$ is the angular velocity. The conservation of angular momentum implies that:
$$
[
L = \text{constant}
]
$$
**Step 4: Expressing Velocity**

The total velocity \( v \) in polar coordinates can be decomposed as:
$$
[
v = \sqrt{\dot{r}^2 + (r \dot{\theta})^2}
]
$$

Substituting $( \dot{\theta} = \frac{L}{\mu r^2} )$ gives:
$$
[
v = \sqrt{\dot{r}^2 + \left( \frac{L}{\mu r} \right)^2}
]
$$
**Step 5: Deriving Velocity at Perihelion**

At perihelion, the distance \( r \) is at its minimum, denoted as $( r_p )$. The radial velocity $( \dot{r} )$ is zero since it is at the closest approach, thus:
$$
[
v_p = \sqrt{0 + \left( \frac{L}{\mu r_p} \right)^2} = \frac{L}{\mu r_p}
]
$$
**Step 6: Analysis of Velocity**

To prove that this velocity is maximum, consider the energy conservation equation at perihelion:
$$
[
E = \frac{1}{2} \mu v_p^2 - \frac{k}{r_p}
]
$$

Since \( E \) is constant and negative (for bound orbits), rearranging gives:
$$
[
\frac{1}{2} \mu v_p^2 = E + \frac{k}{r_p}
]
$$

**Step 7: Velocity Comparison**

As \( r \) increases (moving away from perihelion), \( v \) must decrease due to the balance of energy terms. Hence, at perihelion $( v_p )$ is maximized:

$$
[
\text{If } r > r_p, \quad v < v_p
]
$$

**Conclusion**

Thus, we have shown that the velocity of the reduced mass is maximum at perihelion:
$$
[
v_p = \frac{L}{\mu r_p}
]
$$

## **Question No.6 to 7**
### **Solution**

**Transformation of Coordinates and Calculation of Orbital Parameters**

**Step 1: Orbital Plane Coordinates**

The position vector $(r(\theta))$ in the orbital plane is given by the equation:
$$
[
r(\theta) = \frac{a(1 - e^2)}{1 + e \cos f}
]
$$
where:
- \(a\) is the semi-major axis,
- \(e\) is the eccentricity,
- \(f\) is the true anomaly.

The position in the orbital plane can be expressed in Cartesian coordinates \$((r_x, r_y)\)$ as:
$$
[
r_x = r \cos f, \quad r_y = r \sin f, \quad r_z = 0
]
$$

**Step 2: Transformation from Orbital Plane to Geocentric Coordinates**

To transform the coordinates from the orbital plane to the geocentric frame, we perform three successive rotations:

1. **Rotation by $(\Omega)$** (longitude of the ascending node) around the \(z\)-axis.
2. **Rotation by \(I\)** (inclination) around the \(x\)-axis.
3. **Rotation by $(\omega)$** (argument of pericenter) around the \(z\)-axis.

The transformation matrix $(\mathbf{T})$ is given by:
$$
[
\mathbf{T} =
\begin{pmatrix}
\cos \Omega \cos (\omega + f) - \sin \Omega \sin (\omega + f) \cos I & -\cos \Omega \sin (\omega + f) - \sin \Omega \cos (\omega + f) \cos I & \sin (\omega + f) \sin I \\
\sin \Omega \cos (\omega + f) + \cos \Omega \sin (\omega + f) \cos I & -\sin \Omega \sin (\omega + f) + \cos \Omega \cos (\omega + f) \cos I & -\sin (\omega + f) \sin I \\
\sin I \sin (\omega + f) & \sin I \cos (\omega + f) & \cos I
\end{pmatrix}
]
$$

Thus, the geocentric coordinates \((X, Y, Z)\) are related to the position in the orbital plane as:
$$
[
\begin{pmatrix}
X \\
Y \\
Z
\end{pmatrix}
= r
\begin{pmatrix}
\cos \Omega \cos (\omega + f) - \sin \Omega \sin (\omega + f) \cos I \\
\sin \Omega \cos (\omega + f) + \cos \Omega \sin (\omega + f) \cos I \\
\sin (\omega + f) \sin I
\end{pmatrix}
]
$$

This describes the transformation of the position vector from the orbital plane to the geocentric system.

**Step 3: Compute the Orbital Parameters**

We now compute the semi-major axis \(a\), eccentricity \(e\), and inclination \(I\) in terms of the observed position \((X, Y, Z)\) and velocity $((\dot{X}, \dot{Y}, \dot{Z})).$

**Semi-major axis \(a\):**

The semi-major axis \(a\) can be found using the vis-viva equation:
$$
[
v^2 = \mu \left( \frac{2}{r} - \frac{1}{a} \right)
]
$$

where $(v = \sqrt{\dot{X}^2 + \dot{Y}^2 + \dot{Z}^2})$ is the magnitude of the velocity and $(r = \sqrt{X^2 + Y^2 + Z^2})$ is the radial distance.

Solving for \(a\), we get:
$$
[
a = \frac{1}{\frac{2}{r} - \frac{v^2}{\mu}}
]
$$

**Eccentricity \(e\):**

The eccentricity \(e\) is related to the angular momentum and the energy of the orbit. First, we compute the specific angular momentum vector $(\mathbf{h} = \mathbf{r} \times \mathbf{v}):$
$$
[
\mathbf{h} =
\begin{vmatrix}
\hat{i} & \hat{j} & \hat{k} \\
X & Y & Z \\
\dot{X} & \dot{Y} & \dot{Z}
\end{vmatrix}
]
$$
The magnitude of the angular momentum is:
$$
[
h = \sqrt{(Y \dot{Z} - Z \dot{Y})^2 + (Z \dot{X} - X \dot{Z})^2 + (X \dot{Y} - Y \dot{X})^2}
]
$$

The eccentricity is given by:
$$
[
e = \sqrt{1 - \frac{h^2}{a \mu}}
]
$$

where $(\mu)$ is the gravitational parameter.

**Inclination \(I\):**

The inclination \(I\) is determined from the \(z\)-component of the angular momentum vector. Specifically, we calculate:
$$
[
\cos I = \frac{h_z}{h}
]
$$
where $(h_z = X \dot{Y} - Y \dot{X})$ is the \(z\)-component of the angular momentum.

Thus, the inclination is given by:
$$
[
I = \cos^{-1} \left( \frac{h_z}{h} \right)
]
$$
**Step 4: Conclusion**

We have derived the transformation from the orbital plane to the geocentric system and calculated the orbital parameters \(a\), \(e\), and \(I\) from the observed position \((X, Y, Z)\) and velocity $((\dot{X}, \dot{Y}, \dot{Z}))$.

$$

$$

```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
# Constants
mu = 1
C_tilde = 1  # Dimensionless C_tilde
def equations(t, y):
    r, vr, theta, vtheta = y
    drdt = vr
    dvrdt = r * vtheta**2 - C_tilde / r**2
    dthetadt = vtheta
    dvthetadt = -2 * vr * vtheta / r
    return [drdt, dvrdt, dthetadt, dvthetadt]
initial_conditions = [
    [1, 0.5, np.pi/2, 1],
    [1, 1, np.pi/2, 1],
    [1, 1.03, np.pi/2, 1]
]

t_span = (0, 10)

for i, y0 in enumerate(initial_conditions):
    sol = solve_ivp(equations, t_span, y0, dense_output=True, method='RK45')
    t_eval = np.linspace(t_span[0], t_span[1], 50)
    r_sol, vr_sol, theta_sol, vtheta_sol = sol.sol(t_eval)
    x_sol = r_sol * np.cos(theta_sol)
    y_sol = r_sol * np.sin(theta_sol)
    plt.figure(figsize=(8, 6))
    plt.plot(x_sol, y_sol, label=f'Case {chr(ord("a") + i)}')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title(f'Orbit Trajectory - Case {chr(ord("a") + i)}')
    plt.legend()
    plt.grid(True)
    plt.axis('equal')
    plt.show()
    E_tilde = 0.5 * (vr_sol[0]**2 + r_sol[0]**2 * vtheta_sol[0]**2) - C_tilde / r_sol[0]
    l_tilde = r_sol[0]**2 * vtheta_sol[0]

    print(f"Case {chr(ord('a') + i)}:")
    print(f"  E_tilde: {E_tilde}")
    print(f"  l_tilde: {l_tilde}")
```

```python
# Constants
G = 6.67e-11  # Nm^2/kg^2
me = 6e24     # kg
msun = 2e30    # kg
Rs = 1.5e11    # m (1 AU)
ts = 3.15e7    # s (1 year)

# Reduced mass
mu = me * msun / (me + msun)
C = G * msun * mu
C_tilde = C * ts**2 / (Rs**3)
print(f"C_tilde: {C_tilde}")
r_perihelion = 0.98 * Rs  # m
v_perihelion = 30.28e3    # m/s
r_tilde_perihelion = r_perihelion / Rs
v_tilde_perihelion = v_perihelion * ts / Rs
E_tilde = 0.5 * v_tilde_perihelion**2 - C_tilde / r_tilde_perihelion
l_tilde = r_tilde_perihelion * v_tilde_perihelion

# Eccentricity
eccentricity = np.sqrt(1 + 2 * E_tilde * l_tilde**2 / (C_tilde**2))
print(f"E_tilde: {E_tilde}")
print(f"l_tilde: {l_tilde}")
print(f"Eccentricity: {eccentricity}")
```
