## **Theorem 1. : $$(AB)C = A(BC)$$**
**Proof**
$$
%%latex
\textbf{Proof of the Associative Law of Matrix Multiplication:} \\
\text{Let given three matrices } A \in \mathbb{R}^{m \times n}, B \in \mathbb{R}^{n \times p}, \text{ and } C \in \mathbb{R}^{p \times q}, \text{ we want to prove that } (AB)C = A(BC). \\
\\
$$
**Step 1: Define Matrix Multiplication**

\
$$
\text{Let } A = [a_{ij}] \text{ be an } m \times n \text{ matrix}, B = [b_{jk}] \text{ be an } n \times p \text{ matrix, and } C = [c_{kl}] \text{ be a } p \times q \text{ matrix.} \\
\text{The product of two matrices } AB \text{ is defined as: } \\
(AB)_{ik} = \sum_{j=1}^{n} a_{ij} b_{jk}, \quad \text{for } i = 1, 2, \dots, m \text{ and } k = 1, 2, \dots, p. \\
\\
$$
**Step 2: Compute } (AB)C**

\
$$
\text{Let } D = AB \text{ be the product of matrices } A \text{ and } B. \text{ Then, } D \text{ is an } m \times p \text{ matrix, and } \\
D_{ik} = \sum_{j=1}^{n} a_{ij} b_{jk}. \\
\\
\text{Now, consider the product } (AB)C = DC. \text{ The element at position } (i,l) \text{ in } (AB)C \text{ is:} \\
((AB)C)_{il} = \sum_{k=1}^{p} D_{ik} c_{kl} = \sum_{k=1}^{p} \left( \sum_{j=1}^{n} a_{ij} b_{jk} \right) c_{kl}. \\
\text{This can be expanded as:} \\
((AB)C)_{il} = \sum_{k=1}^{p} \sum_{j=1}^{n} a_{ij} b_{jk} c_{kl}. \quad \cdots (1) \\
\\
$$
**Step 3: Compute } A(BC)**
\
$$
\text{Next, consider the product } BC \text{ first. Let } E = BC \text{ be the product of matrices } B \text{ and } C. \text{ Then, } E \text{ is an } n \times q \text{ matrix, and } \\
E_{jl} = \sum_{k=1}^{p} b_{jk} c_{kl}. \\
\\
\text{Now, consider the product } A(BC) = AE. \text{ The element at position } (i,l) \text{ in } A(BC) \text{ is:} \\
(A(BC))_{il} = \sum_{j=1}^{n} a_{ij} E_{jl} = \sum_{j=1}^{n} a_{ij} \left( \sum_{k=1}^{p} b_{jk} c_{kl} \right). \\
\text{This can be expanded as:} \\
(A(BC))_{il} = \sum_{j=1}^{n} \sum_{k=1}^{p} a_{ij} b_{jk} c_{kl}. \quad \cdots (2) \\
\\
$$

**Step 4: Compare  $$(AB)C \text{ and } A(BC)$$**
$$
\text{By comparing equations } (1) \text{ and } (2), \text{ we observe that:} \\
((AB)C)_{il} = \sum_{k=1}^{p} \sum_{j=1}^{n} a_{ij} b_{jk} c_{kl} = \sum_{j=1}^{n} \sum_{k=1}^{p} a_{ij} b_{jk} c_{kl} = (A(BC))_{il}. \\
\\
\text{Since the elements at each position } (i,l) \text{ are equal for } (AB)C \text{ and } A(BC), \text{ we conclude that:} \\
(AB)C = A(BC). \quad \text{(Associative Law of Matrix Multiplication)} \\
\\
\boxed{\text{Therefore, the associative property holds true: } (AB)C = A(BC).}
$$

## **Theorem 2. : $$(AB)^T = B^T A^T$$**
**Proof**
$$
%%latex
\textbf{Proof of the Transpose of a Product of Matrices:} \\
\text{Let given two matrices } A \in \mathbb{R}^{m \times n} \text{ and } B \in \mathbb{R}^{n \times p}, \text{ we want to prove that } (AB)^T = B^T A^T. \\
$$
**Step 1: Define the Product of Two Matrices**
$$
\text{Let } A = [a_{ij}] \text{ be an } m \times n \text{ matrix and } B = [b_{jk}] \text{ be an } n \times p \text{ matrix. The product } AB \text{ is an } m \times p \text{ matrix defined as:} \\
(AB)_{ik} = \sum_{j=1}^{n} a_{ij} b_{jk}, \quad \text{for } i = 1, 2, \dots, m \text{ and } k = 1, 2, \dots, p. \\
\\
$$
**Step 2: Define the Transpose of a Matrix**
$$
\text{The transpose of a matrix } C = [c_{ij}] \text{ is defined as the matrix } C^T = [c_{ji}], \text{ where rows and columns are swapped.} \\
\\
$$
**Step 3: Compute the Transpose of the Product (AB)**
$$
\text{Let } D = AB \text{ be the product of matrices } A \text{ and } B. \text{ Then, } D \text{ is an } m \times p \text{ matrix, and } \\
D_{ik} = \sum_{j=1}^{n} a_{ij} b_{jk}. \\
\\
\text{The transpose of } D \text{ is } D^T = (AB)^T \text{ and is a } p \times m \text{ matrix. The element at position } (k,i) \text{ in } (AB)^T \text{ is:} \\
((AB)^T)_{ki} = D_{ik} = \sum_{j=1}^{n} a_{ij} b_{jk}. \quad \cdots (1) \\
\\
$$
**Step 4: Compute the Product of the Transposes } B^T \text{ and } A^T**
$$
\text{Consider the matrices } B^T \text{ and } A^T, \text{ where } B^T \text{ is a } p \times n \text{ matrix and } A^T \text{ is an } n \times m \text{ matrix.} \\
\text{The product } B^T A^T \text{ is a } p \times m \text{ matrix, and the element at position } (k,i) \text{ is:} \\
(B^T A^T)_{ki} = \sum_{j=1}^{n} (B^T)_{kj} (A^T)_{ji}. \\
\\
\text{By the definition of the transpose, we have } (B^T)_{kj} = b_{jk} \text{ and } (A^T)_{ji} = a_{ij}. \text{ Thus,} \\
(B^T A^T)_{ki} = \sum_{j=1}^{n} b_{jk} a_{ij}. \quad \cdots (2) \\
\\
$$
**Step 5: Compare } $$(AB)^T \text{ and } B^T A^T$$**
$$
\text{By comparing equations } (1) \text{ and } (2), \text{ we observe that:} \\
((AB)^T)_{ki} = \sum_{j=1}^{n} a_{ij} b_{jk} = \sum_{j=1}^{n} b_{jk} a_{ij} = (B^T A^T)_{ki}. \\
\\
\text{Since the elements at each position } (k,i) \text{ are equal for } (AB)^T \text{ and } B^T A^T, \text{ we conclude that:} \\
(AB)^T = B^T A^T. \quad \text{(Transpose of a Product of Matrices)} \\
\\
\boxed{\text{Therefore, the transpose of a product of two matrices is given by } (AB)^T = B^T A^T.}
$$

## **Theorem 3. :**
$$
{det}(AB) = \text{det}(BA) = \text{det}(A) \times \text{det}(B)
$$

**Step 1: Basic Properties of Determinants**
$$
\text{Let } A \text{ and } B \text{ be square matrices of order } n \text{ The determinant function has the following properties:} \\
1. \ \ {det}(I) = 1, \quad \text{where } I \text{ is the identity matrix.} \\
2. \ \text{det}(AB) = \text{det}(A) \times \text{det}(B). \quad \text{(Multiplicative Property)} \\
$$
**Step 2: Prove**
$$
\text{det}(AB) = \text{det}(BA)
\\
\text{We begin by observing that for square matrices } A \text{ and } B: \\
AB \text{ and } BA \text{ are also square matrices of the same order.} \\
\\
\text{Assume } A \text{ is invertible. Then, we can write:} \\
\text{det}(AB) = \text{det}(A) \times \text{det}(B). \quad \text{(By the multiplicative property of determinants)} \\
\\
\text{Now, consider } \text{det}(BA): \\
\text{det}(BA) = \text{det}(B) \times \text{det}(A). \quad \text{(Again, by the multiplicative property of determinants)} \\
\\
\text{Therefore, } \text{det}(AB) = \text{det}(A) \times \text{det}(B) \quad \text{and} \quad \text{det}(BA) = \text{det}(B) \times \text{det}(A). \\
\text{Since multiplication of scalars is commutative:} \\
\text{det}(A) \times \text{det}(B) = \text{det}(B) \times \text{det}(A). \\
\text{Thus, } \text{det}(AB) = \text{det}(BA). \quad \cdots (1) \\
$$
**Step 3: Consider the General Case**
$$
\text{Even if } A \text{ is not invertible, we can approach the problem using continuity and properties of determinants.} \\
\text{Let } A_{\epsilon} = A + \epsilon I, \text{ where } \epsilon \text{ is a small scalar and } I \text{ is the identity matrix.} \\
\text{For sufficiently small } \epsilon, \text{ the matrix } A_{\epsilon} \text{ is invertible.} \\
\\
\text{Now, we have:} \\
\text{det}(A_{\epsilon} B) = \text{det}(A_{\epsilon}) \times \text{det}(B), \quad \text{and} \\
\text{det}(BA_{\epsilon}) = \text{det}(B) \times \text{det}(A_{\epsilon}). \\
\text{Taking the limit as } \epsilon \to 0, \text{ we get:} \\
\text{det}(AB) = \lim_{\epsilon \to 0} \text{det}(A_{\epsilon} B) = \lim_{\epsilon \to 0} (\text{det}(A_{\epsilon}) \times \text{det}(B)) = \text{det}(A) \times \text{det}(B). \\
\text{Similarly, } \text{det}(BA) = \lim_{\epsilon \to 0} \text{det}(BA_{\epsilon}) = \lim_{\epsilon \to 0} (\text{det}(B) \times \text{det}(A_{\epsilon})) = \text{det}(B) \times \text{det}(A). \\
\text{Hence, } \text{det}(AB) = \text{det}(BA) = \text{det}(A) \times \text{det}(B). \quad \cdots (2) \\
$$
**Step 4: Conclusion**
$$
\text{From equations } (1) \text{ and } (2), \text{ we conclude that:} \\
\boxed{\text{det}(AB) = \text{det}(BA) = \text{det}(A) \times \text{det}(B).}
$$


## **Theorem 4. : $$(AB)^\dagger = B^\dagger A^\dagger$$**

**Proof**

$$
(AB)^\dagger = B^\dagger A^\dagger \textbf{ (Conjugate Transpose of a Product of Matrices)}
$$
**Step 1: Definition of the Conjugate Transpose (Hermitian Transpose)**

$$
\text{For a matrix } A \in \mathbb{C}^{m \times n}, \text{ the conjugate transpose of } A \text{, denoted as } A^\dagger, \text{ is defined as:} \\
A^\dagger = (\overline{A})^T, \quad \text{where } \overline{A} \text{ is the matrix with each entry being the complex conjugate of } A. \\
\text{Thus, } (A^\dagger)_{ij} = \overline{A_{ji}}. \\
\\
$$
**Step 2: Definition of the Product of Two Matrices**

$$
\text{Let } A \in \mathbb{C}^{m \times n} \text{ and } B \in \mathbb{C}^{n \times p}. \\
\text{The product } AB \text{ is an } m \times p \text{ matrix, and its element at position } (i,k) \text{ is:} \\
(AB)_{ik} = \sum_{j=1}^{n} A_{ij} B_{jk}. \quad \cdots (1) \\
\\
$$
**Step 3: Conjugate Transpose of the Product } AB**
$$
\text{We need to find the conjugate transpose of the product } AB. \\
\text{By definition, } (AB)^\dagger = (\overline{AB})^T, \text{ where } \overline{AB} \text{ is the matrix obtained by taking the complex conjugate of each entry in } AB. \\
\\
\text{Taking the complex conjugate of } (1), \text{ we get:} \\
\overline{(AB)_{ik}} = \overline{\sum_{j=1}^{n} A_{ij} B_{jk}}. \\
\text{Since the complex conjugate of a sum is the sum of the complex conjugates, we have:} \\
\overline{(AB)_{ik}} = \sum_{j=1}^{n} \overline{A_{ij}} \, \overline{B_{jk}}. \quad \cdots (2) \\
\\
$$
**Step 4: Transpose of the Conjugate}**
$$
\text{Now, taking the transpose of the matrix with elements given by } (2), \text{ we obtain the element at position } (k,i) \text{ of } (AB)^\dagger: \\
((AB)^\dagger)_{ki} = \overline{(AB)_{ik}} = \sum_{j=1}^{n} \overline{B_{jk}} \, \overline{A_{ij}}. \quad \cdots (3) \\
\\
$$
**Step 5: Conjugate Transpose of } B \text{ and } A**
$$
\text{Now consider the conjugate transposes } B^\dagger \text{ and } A^\dagger: \\
(B^\dagger)_{kj} = \overline{B_{jk}} \quad \text{and} \quad (A^\dagger)_{ji} = \overline{A_{ij}}. \\
\\
\text{Therefore, the product } B^\dagger A^\dagger \text{ is a matrix whose } (k,i) \text{ entry is given by:} \\
(B^\dagger A^\dagger)_{ki} = \sum_{j=1}^{n} (B^\dagger)_{kj} (A^\dagger)_{ji} = \sum_{j=1}^{n} \overline{B_{jk}} \, \overline{A_{ij}}. \quad \cdots (4) \\
\\
$$
**Step 6: Compare the Results}**
$$
\text{Comparing equations } (3) \text{ and } (4), \text{ we see that:} \\
((AB)^\dagger)_{ki} = (B^\dagger A^\dagger)_{ki}. \\
\\
\text{Since the elements at each position } (k,i) \text{ are equal, we conclude that:} \\
\boxed{(AB)^\dagger = B^\dagger A^\dagger}. \\
\\
$$

## **Theorem 5. :**
$$\text{trace}(ABC) = \text{trace}(CAB) = \text{trace}(BCA) \neq \text{trace}(CBA)$$

**Proof**

**Step 1: Definition of Trace**
$$
\text{For a square matrix } A \in \mathbb{C}^{n \times n}, \text{ the trace of } A \text{, denoted by } \text{trace}(A), \text{ is defined as the sum of the diagonal elements:} \\
\text{trace}(A) = \sum_{i=1}^{n} A_{ii}. \\
\\
$$
**Step 2: Trace of a Product of Matrices**
$$
\text{Consider three matrices } A, B, C \in \mathbb{C}^{n \times n}. \\
\text{The } (i,i)\text{th element of the matrix product } ABC \text{ is:} \\
(ABC)_{ii} = \sum_{j=1}^{n} \sum_{k=1}^{n} A_{ij} B_{jk} C_{ki}. \quad \cdots (1) \\
\text{Therefore, the trace of } ABC \text{ is:} \\
\text{trace}(ABC) = \sum_{i=1}^{n} (ABC)_{ii} = \sum_{i=1}^{n} \sum_{j=1}^{n} \sum_{k=1}^{n} A_{ij} B_{jk} C_{ki}. \quad \cdots (2) \\
\\
$$
**Step 3: Cyclic Property of Trace**
$$
\text{Now, consider the matrix product } CAB. \\
\text{The } (i,i)\text{th element of } CAB \text{ is:} \\
(CAB)_{ii} = \sum_{j=1}^{n} \sum_{k=1}^{n} C_{ij} A_{jk} B_{ki}. \quad \cdots (3) \\
$$
$$
\text{Taking the trace, we have:} \\
\text{trace}(CAB) = \sum_{i=1}^{n} (CAB)_{ii} = \sum_{i=1}^{n} \sum_{j=1}^{n} \sum_{k=1}^{n} C_{ij} A_{jk} B_{ki}. \quad \cdots (4) \\
\\
$$

**Step 4: Apply the Cyclic Property Again**
$$
\text{Similarly, consider the matrix product } BCA. \\
\text{The } (i,i)\text{th element of } BCA \text{ is:} \\
(BCA)_{ii} = \sum_{j=1}^{n} \sum_{k=1}^{n} B_{ij} C_{jk} A_{ki}. \quad \cdots (5) \\
$$
$$
\text{Taking the trace, we have:} \\
\text{trace}(BCA) = \sum_{i=1}^{n} (BCA)_{ii} = \sum_{i=1}^{n} \sum_{j=1}^{n} \sum_{k=1}^{n} B_{ij} C_{jk} A_{ki}. \quad \cdots (6) \\
\\
$$

**Step 5: Inequality for Non-Cyclic Permutations**
$$
\text{Now, consider the matrix product } CBA. \\
\text{The } (i,i)\text{th element of } CBA \text{ is:} \\
(CBA)_{ii} = \sum_{j=1}^{n} \sum_{k=1}^{n} C_{ij} B_{jk} A_{ki}. \quad \cdots (7) \\
\text{Taking the trace, we have:} \\
\text{trace}(CBA) = \sum_{i=1}^{n} (CBA)_{ii} = \sum_{i=1}^{n} \sum_{j=1}^{n} \sum_{k=1}^{n} C_{ij} B_{jk} A_{ki}. \quad \cdots (8) \\
\\
\text{Comparing equations } (2), (4), (6), \text{ and } (8), \text{ we see that:} \\
\text{trace}(ABC) = \text{trace}(CAB) = \text{trace}(BCA) \neq \text{trace}(CBA). \\
\\
\textbf{Conclusion:} \\
\text{The trace of a product of matrices is invariant under cyclic permutations, but it does not remain the same under arbitrary permutations.} \\
\boxed{\text{trace}(ABC) = \text{trace}(CAB) = \text{trace}(BCA) \neq \text{trace}(CBA)}
$$

## **Theorem 6. :** Prove All eigenvalues of a symmetric matrix are real numbers.

**Proof**

$$
%%latex
\textbf{All Eigenvalues of a Symmetric Matrix Are Real} \\
\\
$$
**Step 1: Definition of Eigenvalues and Eigenvectors**
$$
\text{Let } A \text{ be an } n \times n \text{ symmetric matrix.} \\
\text{If } \lambda \text{ is an eigenvalue of } A \text{ and } \mathbf{v} \text{ is the corresponding eigenvector, then:} \\
A\mathbf{v} = \lambda \mathbf{v}, \quad \text{where } \mathbf{v} \neq \mathbf{0}. \quad \cdots (1) \\
\\
$$
**Step 2: Taking the Conjugate Transpose**
$$
\text{Take the conjugate transpose of both sides of equation } (1): \\
\mathbf{v}^H A^H = \lambda^* \mathbf{v}^H, \quad \cdots (2) \\
\text{where } \mathbf{v}^H \text{ denotes the conjugate transpose (Hermitian transpose) of } \mathbf{v}, \\
\text{and } \lambda^* \text{ is the complex conjugate of } \lambda. \\
\\
$$
**Step 3: Symmetric Property of Matrix A**
$$
\text{Since } A \text{ is symmetric, } A^H = A. \\
\text{Substitute this into equation } (2): \\
\mathbf{v}^H A = \lambda^* \mathbf{v}^H. \quad \cdots (3) \\
\\
$$
**Step 4: Equating Two Expressions**
$$
\text{From equations } (1) \text{ and } (3), \text{ we have:} \\
\mathbf{v}^H A\mathbf{v} = \lambda \mathbf{v}^H \mathbf{v}, \quad \cdots (4) \\
\mathbf{v}^H A\mathbf{v} = \lambda^* \mathbf{v}^H \mathbf{v}. \quad \cdots (5) \\
\\
$$
**Step 5: Equality of Eigenvalues**
$$
\text{This implies that } \lambda \text{ is a real number, since it is equal to its own complex conjugate.} \\
\\
$$
**Conclusion:** \\
$$
\text{Thus, we have shown that all eigenvalues } \lambda \text{ of a symmetric matrix } A \text{ are real.} \\
\boxed{\text{All eigenvalues of a symmetric matrix are real.}}
$$

## **Theorem 7. :** Prove The determinant of a symmetric matrix A is the product of its eigenvalues.

**Proof**



**Step 1: Symmetric Matrix and Diagonalization** \\
$$
\text{Let } A \text{ be an } n \times n \text{ symmetric matrix.} \\
\text{Since } A \text{ is symmetric, it is diagonalizable. This means there exists an orthogonal matrix } Q \\
\text{(i.e., } Q^T Q = I\text{, where } I \text{ is the identity matrix) and a diagonal matrix } \Lambda \text{ such that:} \\
A = Q \Lambda Q^T. \quad \cdots (1) \\
\text{Here, } \Lambda \text{ is a diagonal matrix whose diagonal elements are the eigenvalues of } A. \\
\\
$$
**Step 2: Determinant of Matrix A** \\
$$
\text{The determinant of } A \text{ is given by:} \\
\text{det}(A) = \text{det}(Q \Lambda Q^T). \quad \cdots (2) \\
\\
$$
**Step 3: Property of Determinants under Similarity Transformations**\\
$$
\text{The determinant of a matrix is invariant under similarity transformations. Specifically,} \\
\text{if } B = P C P^{-1} \text{, then } \text{det}(B) = \text{det}(C). \\
\text{Since } Q \text{ is orthogonal (i.e., } Q^{-1} = Q^T \text{), we have:} \\
\text{det}(A) = \text{det}(Q \Lambda Q^T) = \text{det}(Q) \text{det}(\Lambda) \text{det}(Q^T). \quad \cdots (3) \\
\\
$$
**Step 4: Determinant of Orthogonal Matrices** \\
$$
\text{The determinant of an orthogonal matrix is either } +1 \text{ or } -1. \\
\text{Since } Q^T \text{ is also orthogonal, we have:} \\
\text{det}(A) = \text{det}(Q) \text{det}(\Lambda) \text{det}(Q^T). \\
\text{But } \text{det}(Q^T) = \text{det}(Q) \text{, so:} \\
\text{det}(A) = \text{det}(Q) \text{det}(\Lambda) \text{det}(Q) = \text{det}(Q)^2 \text{det}(\Lambda). \quad \cdots (4) \\
\\
$$
**Step 5: Simplification Using } \text{det}(Q)^2**
$$
\text{Since } \text{det}(Q)^2 = 1 \text{ (because } \text{det}(Q) \text{ is } \pm 1\text{), we have:} \\
\text{det}(A) = 1 \times \text{det}(\Lambda) = \text{det}(\Lambda). \quad \cdots (5) \\
\\
$$
**Step 6: Determinant of a Diagonal Matrix**
$$
\text{The determinant of a diagonal matrix } \Lambda \text{ is the product of its diagonal elements.} \\
\text{Since the diagonal elements of } \Lambda \text{ are the eigenvalues of } A \text{, we have:} \\
\text{det}(A) = \lambda_1 \lambda_2 \dots \lambda_n, \\
\text{where } \lambda_1, \lambda_2, \dots, \lambda_n \text{ are the eigenvalues of } A. \quad \cdots (6) \\
\\
$$
**Conclusion:**
$$
\text{Thus, the determinant of a symmetric matrix } A \text{ is equal to the product of its eigenvalues.} \\
\boxed{\text{det}(A) = \prod_{i=1}^{n} \lambda_i.}
$$

## **Theorem 8. :** Prove Every symmetric matrix can be diagonalized by an orthogonal matrix.

**Proof**

**Step 1: Symmetric Matrix and Eigenvalues** \\
$$
\text{Let } A \text{ be an } n \times n \text{ symmetric matrix, i.e., } A = A^T. \\
\text{We want to show that } A \text{ can be diagonalized by an orthogonal matrix.} \\
\text{Specifically, we aim to find an orthogonal matrix } Q \text{ such that:} \\
A = Q \Lambda Q^T, \\
\text{where } \Lambda \text{ is a diagonal matrix containing the eigenvalues of } A. \\
\\
$$
**Step 2: Existence of Real Eigenvalues** \\
$$
\text{Since } A \text{ is symmetric, all its eigenvalues are real.} \\
\text{Let } \lambda_1, \lambda_2, \dots, \lambda_n \text{ be the eigenvalues of } A, \text{ and let } \mathbf{v}_1, \mathbf{v}_2, \dots, \mathbf{v}_n \text{ be the corresponding eigenvectors.} \\
\\
$$
**Step 3: Orthogonality of Eigenvectors** \\
$$
\text{For a symmetric matrix } A, \text{ the eigenvectors corresponding to distinct eigenvalues are orthogonal.} \\
\text{This means } \mathbf{v}_i^T \mathbf{v}_j = 0 \text{ for } i \neq j. \quad \cdots (1) \\
\\
$$

**Step 4: Constructing the Orthogonal Matrix  Q** \\

$$
\text{We can normalize the eigenvectors to obtain an orthonormal set.} \\
\text{Let } \mathbf{u}_i = \frac{\mathbf{v}_i}{\|\mathbf{v}_i\|}, \text{ where } \|\mathbf{v}_i\| \text{ is the norm of } \mathbf{v}_i. \\
\text{Then } \mathbf{u}_1, \mathbf{u}_2, \dots, \mathbf{u}_n \text{ are orthonormal eigenvectors of } A. \\
\text{Construct the orthogonal matrix } Q \text{ as:} \\
Q = \begin{pmatrix} \mathbf{u}_1 & \mathbf{u}_2 & \dots & \mathbf{u}_n \end{pmatrix}. \quad \cdots (2) \\

## **Theorem 10. :** Every Hermitian matrix can be diagonalized by a unitary matrix.

**Proof**


**Step 1: Definition of a Hermitian Matrix** \\
$$
\text{Let } A \text{ be an } n \times n \text{ Hermitian matrix, i.e., } A = A^\dagger, \text{ where } A^\dagger \text{ is the conjugate transpose of } A. \\
\text{We want to show that } A \text{ can be diagonalized by a unitary matrix } U. \\
\\
$$
**Step 2: Existence of Eigenvalues and Eigenvectors** \\
$$
\text{Since } A \text{ is Hermitian, it has } n \text{ eigenvalues (not necessarily distinct), and all eigenvalues are real.} \\
\text{Let } \lambda_1, \lambda_2, \dots, \lambda_n \text{ be the eigenvalues of } A, \text{ and let } \mathbf{v}_1, \mathbf{v}_2, \dots, \mathbf{v}_n \text{ be the corresponding eigenvectors.} \\
\\
$$
**Step 3: Orthogonality of Eigenvectors**\\
$$
\text{For a Hermitian matrix, the eigenvectors corresponding to distinct eigenvalues are orthogonal.} \\
\text{Even if some eigenvalues are repeated, we can apply the Gram-Schmidt process to obtain an orthogonal basis of eigenvectors.} \\
\text{Thus, we can assume that } \mathbf{v}_1, \mathbf{v}_2, \dots, \mathbf{v}_n \text{ form an orthogonal set of eigenvectors.} \\
\\
$$
**Step 4: Normalization of Eigenvectors** \\
$$
\text{We can normalize the eigenvectors to make them orthonormal. Let } \mathbf{u}_i = \frac{\mathbf{v}_i}{\|\mathbf{v}_i\|} \text{ for } i = 1, 2, \dots, n. \\
\text{Then, the set } \mathbf{u}_1, \mathbf{u}_2, \dots, \mathbf{u}_n \text{ is an orthonormal set of eigenvectors.} \\
\\
$$
**Step 5: Construct the Unitary Matrix } U** \\
$$
\text{Form the matrix } U \text{ whose columns are the orthonormal eigenvectors } \mathbf{u}_1, \mathbf{u}_2, \dots, \mathbf{u}_n: \\
U = \begin{bmatrix} \mathbf{u}_1 & \mathbf{u}_2 & \dots & \mathbf{u}_n \end{bmatrix}. \\
\text{Since the columns of } U \text{ are orthonormal, } U \text{ is a unitary matrix, i.e., } U^\dagger U = I, \text{ where } I \text{ is the identity matrix.} \\
\\
$$
**Step 6: Diagonalization of } A \text{ by } U** \\
$$
\text{Consider the product } U^\dagger A U. \\
\text{Since } A \mathbf{u}_i = \lambda_i \mathbf{u}_i \text{ for each } i, \text{ we have:} \\
U^\dagger A U = \begin{bmatrix} \mathbf{u}_1^\dagger \\ \mathbf{u}_2^\dagger \\ \vdots \\ \mathbf{u}_n^\dagger \end{bmatrix} \begin{bmatrix} A \mathbf{u}_1 & A \mathbf{u}_2 & \dots & A \mathbf{u}_n \end{bmatrix} = \begin{bmatrix} \lambda_1 & 0 & \dots & 0 \\ 0 & \lambda_2 & \dots & 0 \\ \vdots & \vdots & \ddots & \vdots \\ 0 & 0 & \dots & \lambda_n \end{bmatrix}. \\
\text{Thus, } U^\dagger A U \text{ is a diagonal matrix with the eigenvalues } \lambda_1, \lambda_2, \dots, \lambda_n \text{ on the diagonal.} \\
\\
$$
**Conclusion:** \\
$$
\text{We have shown that } A \text{ can be diagonalized by the unitary matrix } U. \\
\boxed{\text{Every Hermitian matrix can be diagonalized by a unitary matrix.}}
$$

### **Understanding the Affine Transformation Matrix in Image Processing**

#### **1. Introduction**

An **affine transformation** is a linear mapping method that preserves points, straight lines, and planes. In image processing, affine transformations are used to correct distortions, such as those caused by the motion of a spacecraft capturing images.

#### **2. The Affine Transformation Matrix**

The affine transformation matrix \( T \) Typically can be represented as:
$$
[T] =
\begin{pmatrix}
a_{11} & a_{12} & t_x \\
a_{21} & a_{22} & t_y \\
0 & 0 & 1
\end{pmatrix}
$$
where:
- $$( a_{11}, a_{12}, a_{21}) and ( a_{22} )$$ represent the linear transformation components, including rotation, scaling, and shearing.
- $$( t_x ) and ( t_y )$$ represent translation along the x and y axes, respectively.

#### **3. Affine Transformation and Its Components**

1. **Translation**:
    - Translation moves the image from one location to another without changing its orientation or scale.
    - Translation is controlled by \( t_x \) and \( t_y \) in the matrix \( T \).

2. **Rotation**:
    - Rotation changes the orientation of the image by rotating it around a fixed point, usually the origin.
    - Rotation is controlled by the elements $$( a_{11}, a_{12}, a_{21}, ) and ( a_{22} )$$.

    For example, a pure rotation by an angle $$( \theta )$$ is represented by:

    \[
    \begin{pmatrix}
    \cos\theta & -\sin\theta & 0 \\
    \sin\theta & \cos\theta & 0 \\
    0 & 0 & 1
    \end{pmatrix}
    \]

3. **Scaling**:
    - Scaling changes the size of the image by expanding or contracting it along the x and y axes.
    - Scaling is also controlled by the elements $$( a_{11}, a_{12}, a_{21}, \) and ( a_{22} )$$.

    For example, a scaling matrix is:

    \[
    \begin{pmatrix}
    s_x & 0 & 0 \\
    0 & s_y & 0 \\
    0 & 0 & 1
    \end{pmatrix}
    \]

    where $$( s_x ) and ( s_y )$$ are the scaling factors along the x and y axes, respectively.

4. **Shearing**:
    - Shearing shifts one part of the image more than another, typically creating a slanted effect.
    - Shearing is determined by off-diagonal elements of the transformation matrix.

    For example, a shear transformation matrix is:

    \[
    \begin{pmatrix}
    1 & k_x & 0 \\
    k_y & 1 & 0 \\
    0 & 0 & 1
    \end{pmatrix}
    \]

    where $$( k_x ) and ( k_y )$$ are shear factors.
