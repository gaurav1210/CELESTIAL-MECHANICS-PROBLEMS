## **Question No.1**

### **1. Solution**
#### **Finding a Normal Vector to the Curve**

**Given:**

The curve equation is
$
[ y^2 = x^3 ]
$
and we need to find a normal vector at the point \( (4, -8) \).

#### **Step 1. Implicit Differentiation**
Differentiate $( y^2 = x^3 )$ implicitly with respect to ( x ):
$$
[
\frac{d}{dx}(y^2) = \frac{d}{dx}(x^3)
]
$$
Applying the chain rule, we get:
$$
[
2y \frac{dy}{dx} = 3x^2
]
$$
Solving for $( \frac{dy}{dx})$ we find:
$$
[
\frac{dy}{dx} = \frac{3x^2}{2y}
]
$$
#### **Step 2. Compute the Tangent Slope at \( (4, -8) \)**

Substitute $$( x = 4 ) and ( y = -8 )$$ into the derivative:
$$
[
\frac{dy}{dx} \bigg|_{(4, -8)} = \frac{3 \cdot 4^2}{2 \cdot (-8)} = \frac{48}{-16} = -3
]
$$
Thus,

The slope of the tangent line at **\( (4, -8) \)** is **\( -3 \).**

#### **Step 3. Compute the Slope of the Normal Line**

The slope of the normal line is the negative reciprocal of the tangent slope:
$$
[
\text{slope of the normal line} = -\frac{1}{-3} = \frac{1}{3}
]
$$
#### **Step 4. Normal Vector**

A normal vector to the curve can be represented by the direction vector of the normal line. For a slope of $( \frac{1}{3})$

The direction vector is:
$
\begin{pmatrix}
1 \\
\frac{1}{3}
\end{pmatrix}
$

To avoid fractions, we can multiply by 3:
$$
[
\mathbf{n} = \begin{pmatrix}
3 \\
1
\end{pmatrix}
]
$$

**Thus, a normal vector to the curve at the point \( (4, -8) \) is:**

$$
[
\mathbf{n} = \begin{pmatrix}
3 \\
1
\end{pmatrix}
]
$$

#### **Finding the Angle Between the Surfaces**

Given the surfaces:
$$
[
      x y^2 z = 3x + z^2 =>  
      f(x, y, z) = x y^2 z - 3x - z^2 = 0

$$

and
$$
[
     3x^2 - y^2 + 2z = 1  =>  g(x, y, z) = 3x^2 - y^2 + 2z - 1 = 0

we want to find the smaller angle between their normals at the point $$((1, -2, 1)).$$

#### **Step 1. Compute the Gradients (Normal Vectors)**

The gradient of a function $$f(x, y, z)$$ is given by:
$$
[
\nabla f = \left( \frac{\partial f}{\partial x}, \frac{\partial f}{\partial y}, \frac{\partial f}{\partial z} \right)
]
$$
###### **Gradient of** $$f(x, y, z) = x y^2 z - 3x - z^2 $$
$$
[
\frac{\partial f}{\partial x} = y^2 z - 3

[
\frac{\partial f}{\partial y} = 2x y z
]
[
\frac{\partial f}{\partial z} = x y^2 - 2z
]
$$
Thus, the gradient vector is:
$$
[
\nabla f = \left( y^2 z - 3, 2x y z, x y^2 - 2z \right)
]
$$
Substituting the point $$((1, -2, 1)) into (\nabla f):$$
$$
[
\nabla f(1, -2, 1) = \left( (-2)^2 \cdot 1 - 3, 2 \cdot 1 \cdot (-2) \cdot 1, 1 \cdot (-2)^2 - 2 \cdot 1 \right)
]
[
\nabla f(1, -2, 1) = \left( 4 - 3, -4, 4 - 2 \right) = (1, -4, 2)
]
$$
###### **Gradient of** $$ g(x, y, z) = 3x^2 - y^2 + 2z - 1:$$
$$
[
\frac{\partial g}{\partial x} = 6x
]
[
\frac{\partial g}{\partial y} = -2y
]
[
\frac{\partial g}{\partial z} = 2
]
$$
Thus, the gradient vector is:
$$
[
\frac{\partial g}{\partial x} = (6x, -2y, 2)
]
$$
Substituting the point $$((1, -2, 1)) into (\nabla g):$$
$$
[
\nabla g(1, -2, 1) = \left( 6 \cdot 1, -2 \cdot (-2), 2 \right)
]
[
\nabla g(1, -2, 1) = (6, 4, 2)
]
$$
#### **Step 2. Find the Angle Between the Normal Vectors**

The angle $( \theta )$ between two vectors $( \mathbf{a}) and ( \mathbf{b} )$ is given by:
$$
[
\cos \theta = \frac{\mathbf{a} \cdot \mathbf{b}}{|\mathbf{a}| |\mathbf{b}|}
]
$$
where $( \mathbf{a} \cdot \mathbf{b})$ is the dot product of the vectors, and $( |\mathbf{a}|) and ( |\mathbf{b}| )$ are their magnitudes.

###### **Vector Field Analysis**

Given the vector field $(\mathbf{F} = 4xy^2z \, \hat{i} - 4/3y^3z \, \hat{j} + x^3y^3 \, \hat{k})$,

1. Show that $(\mathbf{F})$ is solenoidal.

#### **1. Show that (F) is Solenoidal**

A vector field **(F)** is solenoidal if its divergence is zero everywhere.

 i.e. $$( \nabla \cdot \mathbf{F} = 0 ).$$

##### **Compute the Divergence of {F}**

The vector field is:
$$
[
\mathbf{F} = F_1 \, \hat{i} + F_2 \, \hat{j} + F_3 \, \hat{k}
]
$$
where:
$$
[
F_1 = 4xy^2z, \quad F_2 = -4/3y^3z, \quad F_3 = x^3y^3
]
$$
The divergence of **{F}** is given by:
$$
[
\nabla \cdot \mathbf{F} = \frac{\partial F_1}{\partial x} + \frac{\partial F_2}{\partial y} + \frac{\partial F_3}{\partial z}
]
$$
##### **Calculate** $$(\frac{\partial F_1}{\partial x}):$$
$$
[
\frac{\partial F_1}{\partial x} = \frac{\partial}{\partial x}(4xy^2z) = 4y^2z
]
$$
##### **Calculate** $$(\frac{\partial F_2}{\partial y}):$$
$$
[
\frac{\partial F_2}{\partial y} = \frac{\partial}{\partial y}(-4/3y^3z) = -4y^2z
]
$$
##### **Calculate** $$(\frac{\partial F_3}{\partial z}):$$
$$
[
\frac{\partial F_3}{\partial z} = \frac{\partial}{\partial z}(x^3y^3) = 0
]
$$
Thus, the divergence is:
$$
[
\nabla \cdot \mathbf{F} = 4y^2z - 4y^2z + 0 = 0
]
$$
Since,

 The divergence is zero (0).

 **Therefore the following vector field (F) this field is solenoidal.**



#### **2. Find a vector field** $( \mathbf{G} )$ such that $( \mathbf{F} = \nabla \times \mathbf{G} )$


Given the vector field:
$$
[
\mathbf{F} = 4 x y^2 z \hat{i} - \frac{4}{3} y^3 z \hat{j} + x^3 y^3 \hat{k},
]
$$

We want to find a vector field $(\mathbf{G})$ such that:
$$
[
\mathbf{F} = \nabla \times \mathbf{G}.

\nabla \times \mathbf{G} = \mathbf{F}.

$$

3. **For the $(\hat{k})$ component:**
$$
[
\frac{\partial G_y}{\partial x} - \frac{\partial G_x}{\partial y} = x^3 y^3.

Thus:
$$
[
\mathbf{G} = 0 \hat{i} + \left( \frac{x^4 y^3}{4} + C_1 \right) \hat{j} + \left( \frac{4}{3} x y^3 z + C_2 \right) \hat{k}.

$$

To verify, we compute:
$$
[
\nabla \times \mathbf{G}.
]
$$
1. **For the $(\hat{i})$ component:**

[
\frac{\partial G_y}{\partial z} = 0.

[
\frac{\partial G_z}{\partial y} - \frac{\partial G_y}{\partial z} = 4 x y^2 z.

$$

This matches the given vector field $(\mathbf{F})$, verifying that the solution is correct.















## **Question No. 4**
## **Solution**

Given the vector field:
$$
[
\mathbf{V} = (x + 2y + az) \, \hat{i} + (bx - 3y - z) \, \hat{j} + (4x + cy + 2z) \, \hat{k}
]
$$
We need to:

1. Find the constants **\(a\), \(b\), and \(c\)** such that **{V}** is irrotational.
2. Show that **{V}** can be expressed as the gradient of a scalar field $$(\phi(x, y, z)).$$
3. Evaluate the scalar field $$(\phi(x, y, z)).$$

##### **1. Determine Constants \(a\), \(b\), and \(c\)**

For **{V}** to be irrotational, its curl must be zero, i.e., $$( \nabla \times \mathbf{V} = \mathbf{0}).$$

##### **Compute the Curl of {V}**

The curl of **{V}** is given by:
$$
[
\nabla \times \mathbf{V} = \left( \frac{\partial V_3}{\partial y} - \frac{\partial V_2}{\partial z} \right) \hat{i} + \left( \frac{\partial V_1}{\partial z} - \frac{\partial V_3}{\partial x} \right) \hat{j} + \left( \frac{\partial V_2}{\partial x} - \frac{\partial V_1}{\partial y} \right) \hat{k}
]
$$
where $$( V_1 = x + 2y + az ), ( V_2 = bx - 3y - z ), and ( V_3 = 4x + cy + 2z ).
$$
##### **Calculate** $$( \frac{\partial V_3}{\partial y} ) and ( \frac{\partial V_2}{\partial z} ):
[
\frac{\partial V_3}{\partial y} = c, \quad \frac{\partial V_2}{\partial z} = -1
]
$$
So,

The **\(i\)**-component of the curl is:
$$
[
\frac{\partial V_3}{\partial y} - \frac{\partial V_2}{\partial z} = c + 1
]
$$
##### **Calculate** $$( \frac{\partial V_1}{\partial z} ) and ( \frac{\partial V_3}{\partial x} ):$$
$$
[
\frac{\partial V_1}{\partial z} = a, \quad \frac{\partial V_3}{\partial x} = 4
]
$$
So,

The **\(j\)**-component of the curl is:
$$
[
\frac{\partial V_1}{\partial z} - \frac{\partial V_3}{\partial x} = a - 4
]
$$
##### **Calculate** $$( \frac{\partial V_2}{\partial x} ) and ( \frac{\partial V_1}{\partial y} ):$$
$$
[
\frac{\partial V_2}{\partial x} = b, \quad \frac{\partial V_1}{\partial y} = 2
]
$$
So,

The **\(k\)**-component of the curl is:
$$
[
\frac{\partial V_2}{\partial x} - \frac{\partial V_1}{\partial y} = b - 2
]
$$
###### **Set the Curl Components to Zero**

For the vector field **{V}** to be irrotational:
$$
[
c + 1 = 0, \quad a - 4 = 0, \quad b - 2 = 0
]
$$
Thus, we get:
$$
[
a = 4, \quad b = 2, \quad c = -1
]
$$

Therefore,

**(a = 4, b = 2 and c = -1)**


##### **2. Show that** $( \mathbf{V} )$ Can Be Expressed as the Gradient of a Scalar Field $( \phi(x, y, z) )$

To express $( \mathbf{V} )$ as the gradient of a scalar field $( \phi(x, y, z) )$, we must have:

$$[
\mathbf{V} = \nabla \phi = \left(\frac{\partial \phi}{\partial x}\right) \hat{i} + \left(\frac{\partial \phi}{\partial y}\right) \hat{j} + \left(\frac{\partial \phi}{\partial z}\right) \hat{k}
]$$

**Given:**

$$[
\mathbf{V} = (x + 2y + 4z) \, \hat{i} + (2x - 3y - z) \, \hat{j} + (4x - y + 2z) \, \hat{k}

We match the components:
$$
[
\frac{\partial \phi}{\partial x} = x + 2y + 4z
]$$
$$
[
\frac{\partial \phi}{\partial y} = 2x - 3y - z
]
$$
$$
[
\frac{\partial \phi}{\partial z} = 4x - y + 2z
]
$$

##### **3. Evaluate the Scalar Field $( \phi(x, y, z) )$**

We integrate each equation to find $( \phi(x, y, z) ).$

From $( \frac{\partial \phi}{\partial x} = x + 2y + 4z ):$

$$[
\phi(x, y, z) = \frac{x^2}{2} + 2xy + 4xz + f(y, z)
]
$$
where $( f(y, z) )$ is a function of $( y )$ and $( z )$, Since differentiating with respect to $( x )$ should eliminate it.

From $( \frac{\partial \phi}{\partial y} = 2x - 3y - z ):$

Substitute $( \phi(x, y, z) = \frac{x^2}{2} + 2xy + 4xz + f(y, z) )$ into this equation:
$$
[
\frac{\partial}{\partial y} \left( \frac{x^2}{2} + 2xy + 4xz + f(y, z) \right) = 2x + \frac{\partial f(y, z)}{\partial y}
]
$$
This gives:
$$
[
2x - 3y - z = 2x + \frac{\partial f(y, z)}{\partial y}
]
$$
Thus:
$$
[
\frac{\partial f(y, z)}{\partial y} = -3y - z
]
$$
Integrating with respect to $( y ):$
$$
[
f(y, z) = -\frac{3y^2}{2} - yz + g(z)
]
$$
where $( g(z) )$ is a function of $( z ).$

From $( \frac{\partial \phi}{\partial z} = 4x - y + 2z ):$

Substitute $( f(y, z) = -\frac{3y^2}{2} - yz + g(z) )$ into the expression for $( \phi(x, y, z) ):$
$$
[
\phi(x, y, z) = \frac{x^2}{2} + 2xy + 4xz - \frac{3y^2}{2} - yz + g(z)
]
$$
Now differentiate with respect to $( z ):$
$$
[
\frac{\partial}{\partial z} \left( \frac{x^2}{2} + 2xy + 4xz - \frac{3y^2}{2} - yz + g(z) \right) = 4x - y + \frac{dg(z)}{\partial z}
]
$$
This gives:
$$
[
4x - y + 2z = 4x - y + \frac{dg(z)}{\partial z}
]
$$
Thus:
$$
[
\frac{dg(z)}{\partial z} = 2z
]
$$
Integrating with respect to $( z ):$
$$
[
g(z) = z^2 + C
]
$$
##### **Final Expression for** $( \phi(x, y, z) )$

Thus,

The scalar potential $( \phi(x, y, z) )$ is:
$$
[
\phi(x, y, z) = \frac{x^2}{2} + 2xy + 4xz - \frac{3y^2}{2} - yz + z^2 + C
]
$$
Where $( C )$ is an arbitrary constant.



## **Question No. 5**
## **Solution**
#### **Proof: The Path of a Planet Around the Sun is an Ellipse with the Sun at One of the Foci**

##### **Step 1: Gravitational Force and Newton's Second Law**

The gravitational force between the Sun (mass \(M\)) and a planet (mass \(m\)) is given by:
$$
[
\mathbf{F} = -\frac{GMm}{r^2} \hat{\mathbf{r}}
]
$$
where:

- \(G\) is the gravitational constant,
- $(r = |\mathbf{r}|)$ is the distance between the Sun and the planet,
- $(\hat{\mathbf{r}} = \frac{\mathbf{r}}{r})$ is the unit vector along the line joining the Sun and the planet.

According to Newton's second law, the acceleration $(\mathbf{a})$ of the planet is given by:
$$
[
m \mathbf{a} = \mathbf{F}
]
$$
Thus,
$$
[
\mathbf{a} = -\frac{GM}{r^2} \hat{\mathbf{r}}
]
$$
##### **Step 2: Equation of Motion in Polar Coordinates**

The position vector of the planet $(\mathbf{r}(t))$ in polar coordinates is given by:
$$
[
\mathbf{r}(t) = r(t) \hat{\mathbf{r}}(t)
]
$$
The velocity $(\mathbf{v}(t))$ and acceleration $(\mathbf{a}(t))$ are given by:
$$
[
\mathbf{v}(t) = \dot{r} \hat{\mathbf{r}} + r \dot{\theta} \hat{\boldsymbol{\theta}}
]
$$
$$
[
\mathbf{a}(t) = \left(\ddot{r} - r \dot{\theta}^2\right) \hat{\mathbf{r}} + \left(r \ddot{\theta} + 2\dot{r} \dot{\theta}\right) \hat{\boldsymbol{\theta}}
]
$$
where:

- $(\dot{r}) and (\ddot{r})$ are the first and second derivatives of $(r(t))$ with respect to time, $(\dot{\theta}) and (\ddot{\theta})$ are the first and second derivatives of $(\theta(t))$ with respect to time,
- $(\hat{\boldsymbol{\theta}})$ is the unit vector perpendicular to $(\hat{\mathbf{r}})$.

##### **Step 3: Conservation of Angular Momentum**

Angular momentum $(\mathbf{L})$ is conserved in a central force field, meaning:
$$
[
\mathbf{L} = \mathbf{r} \times \mathbf{v} = \text{constant}
]
$$
In scalar form:
$$
[
L = m r^2 \dot{\theta} = \text{constant}
]
$$
Thus, the angular velocity $(\dot{\theta})$ is given by:
$$
[
\dot{\theta} = \frac{L}{m r^2}
]
$$
##### **Step 4: Energy Conservation and the Orbit Equation**

The total mechanical energy \(E\) of the planet is conserved:
$$
[
E = \frac{1}{2} m \mathbf{v}^2 - \frac{GMm}{r} = \text{constant}
]
$$
Substituting the expressions for $(\mathbf{v})$ and simplifying, we obtain:
$$
[
\frac{1}{2} m \left( \dot{r}^2 + \frac{L^2}{m^2 r^2} \right) - \frac{GMm}{r} = \text{constant}
]
$$
Multiplying by 2 and rearranging:
$$
[
\dot{r}^2 + \frac{L^2}{m^2 r^2} = \frac{2E}{m} + \frac{2GM}{r}
]
$$

Let $$(u = \frac{1}{r}),$$ so that $$(\dot{u} = -\frac{\dot{r}}{r^2}) and (\ddot{u} = -\frac{\ddot{r}}{r^2} + 2\frac{\dot{r}^2}{r^3}).$$ The above energy equation becomes:
$$
[
\left(\frac{du}{d\theta}\right)^2 + u^2 = \frac{GMm}{L^2} + \frac{2E}{L^2 u}
]
$$
This differential equation has a solution of the form:
$$
[
u(\theta) = \frac{GMm}{L^2} \left(1 + e \cos\theta\right)
]
$$
where \(e\) is the eccentricity of the orbit.

##### **Step 5: Equation of an Ellipse**

Inverting $(u(\theta))$ to obtain $(r(\theta)):$
$$
[
r(\theta) = \frac{L^2}{GMm \left(1 + e \cos\theta\right)}
]
$$
This is the polar form of an ellipse, confirming that the orbit of the planet is an ellipse.

##### **Conclusion**

The path of a planet around the Sun is indeed an ellipse with the Sun located at one of the foci, as derived using vector calculus and the conservation laws.

```python
import numpy as np                                                              # Importing the numpy library for array manipulation
import matplotlib.pyplot as plt                                                 # Importing the matplotlib library for plotting

G = 6.67430e-11                                                                 # Gravitational constant (m^3 kg^-1 s^-2)
M = 1.989e30                                                                    # Mass of the Sun (kg)
L = 3.2e43                                                                      # Angular momentum (m^2 kg s^-1)
e = 0.5                                                                         # Eccentricity of the orbit (0 <= e < 1)
a = 1                                                                           # Semi-major axis in Astronomical Units (AU) for scaling

p = a * (1 - e**2)                                                              # Semi-latus rectum (p) for the ellipse
def r(theta, e, p):                                                             # Define the polar equation of the ellipse (Kepler's First Law)
    return p / (1 + e * np.cos(theta))
theta = np.linspace(0, 2 * np.pi, 1000)                                         # Generate theta values (0 to 2π)
r_values = r(theta, e, p)                                                       # Calculate the radial distance r for each theta
x = r_values * np.cos(theta)                                                    # Convert from polar to Cartesian coordinates for plotting
y = r_values * np.sin(theta)
plt.figure(figsize=(8, 6))                                                      # Plot the orbit
plt.plot(x, y, label="Planet's Elliptical Orbit", color='b')                    # Plot the elliptical orbit

plt.plot(-e * a, 0, 'ro', label="Sun (Focus)", markersize=12)                   # Plot the Sun at one of the foci
plt.xlim([-1.5*a, 1.5*a])                                                       # Set up plot limits for a better view
plt.ylim([-1.2*a, 1.2*a])
plt.title('Elliptical Orbit of a Planet Around the Sun', fontsize=14)           # Adding labels and title
plt.xlabel('x (AU)', fontsize=12)
plt.ylabel('y (AU)', fontsize=12)
plt.grid(True)
plt.gca().set_aspect('equal')                                                   # Equal scaling for x and y axes
plt.legend()                                                                    # Show legend
plt.show()                                                                      # Display the plot
```

## **Question No. 6**

##### **Calculation of Curl and Line Integral**

**Given the vector field:**
$$
[
\mathbf{F} = \left(-\frac{y}{x^2 + y^2}\right) \hat{i} + \left(\frac{x}{x^2 + y^2}\right) \hat{j}

```python
t = sp.symbols('t')                                                             # Define the parameter t and the parametric equations of r(t)
r = sp.Matrix([t, t**2, t**3])                                                  # r(t) = (t, t^2, t^3)
x, y, z = sp.symbols('x y z')                                                   # Define the vector field F(x, y, z) = (x^2*y, y^2*z, z^2*x)
F = sp.Matrix([x**2 * y, y**2 * z, z**2 * x])
F_at_r_t = F.subs({x: r[0], y: r[1], z: r[2]})                                  # Substitute r(t) = (t, t^2, t^3) into the vector field F
r_prime = r.diff(t)                                                             # Compute the derivative of r(t) with respect to t
dot_product = F_at_r_t.dot(r_prime)                                             # Compute the dot product F(r(t)) · r'(t)
line_integral = sp.integrate(dot_product, (t, 0, 1))                            # Compute the line integral from t = 0 to t = 1
fraction_result = sp.simplify(line_integral)                                    # Output the result as a fraction and as a decimal
decimal_result = line_integral.evalf()
print("line integeral value is :", fraction_result)                             # Printing the line integral value in fraction
print("line integeral value in decimal value is :", decimal_result)             # Printing the line integral value in decimal
```
